# DataFrame.pivot_table(**kwargs)

```python
DataFrame.pivot_table(
    values=None,
    index=None,
    columns=None,
    aggfunc='mean',
    fill_value=None,
    margins=False,
    dropna=True,
    margins_name='All',
)```

````python
#average tip of each gender for each day
tips.pivot_table(values = 'tip', index = 'sex', columns = 'day', aggfunc = 'mean')
````

````python
#total tip from different time on each day
tips.pivot_table('tip', index = 'time', columns = 'day', aggfunc = 'sum', fill_value = 0)
````

````python
#total and average tip for each gender for each day
tips.pivot_table(values = 'tip', index = 'sex', columns = 'day', aggfunc = ['mean', 'sum'])
````

````python
#passing a dict to aggfunc
tips.pivot_table(index = 'sex', columns = 'day', aggfunc = {'tip' : ['mean', 'sum'], 'total_bill': 'sum'})
````

````python
#multiINdex
tips.pivot_table(['tip', 'total_bill'], index = ['sex', 'smoker'], columns = ['day', 'time'])
````
