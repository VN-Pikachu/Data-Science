# DataFrame

<ul>
<li><b><code>iteritems()</code></b>: Iterates over each column as key, value pair with label as key and column value as a Series object</li>
<li><b><code>iterrows()</code></b>: iterrows() returns the iterator yielding each index value along with a series containing the data in each row.</li>
<li><b><code>itertuples</code></b>: return an iterator yielding a named tuple for each row in the DataFrame. </li>
</ul>

```python
pokemon = pd.read_csv('./pokemon.csv', index_col = 0)
pokemon.head(3)
```

```python
#iterate through each column
for column_name, column in pokemon.iteritems():
    print(column_name)
    print(column)
    print('_' * 50)
```

```python
#iterate over each row
for index, row in pokemon.iterrows():
    print(index)
    print(row)
    print('_' * 50)
```

```python
#Iterate over DataFrame rows as namedtuples.
for row in pokemon.itertuples():
    print(row)
```

```python
#iterate over column names:
for column_name in pokemon: #equivalent: for column_name in pokemon.keys()
    print(column_name)
```

```python
#iterate over (col_name, col)
for col_name, col in pokemon.items():
        print(col_name)
        print(col)
        print('-' * 50)
```
