````python
#barh
pokemon.groupby('Generation').size().plot.barh()
````

````python
#bar
pokemon.groupby('Legendary').size().plot.bar(rot = 0)
````

````python
#side by side bar
pokemon.groupby('Legendary')['HP', 'Attack', 'Defense', 'Speed'].mean().plot.bar(rot = True)
````

````python
#stacked bar
pokemon.groupby('Generation')['HP', 'Attack', 'Defense'].mean().plot(kind = 'bar', stacked = True)
````

````python
#plot a hist of each column in each subplot
pokemon.loc[:, 'HP':'Speed'].plot.hist(alpha = .5, subplots = True)
````

````python
pokemon.loc[:, 'HP':'Speed'].plot(kind = 'box')
````
