## how

````python
df6 = pd.DataFrame({'name': ['Peter', 'Paul', 'Mary'],
                    'food': ['fish', 'beans', 'bread']},
                   columns=['name', 'food'])
df6
````

````python
df7 = pd.DataFrame({'name': ['Mary', 'Joseph'],
                    'drink': ['wine', 'beer']},
                   columns=['name', 'drink'])
df7
````

````python
#left join
pd.merge(df6, df7, how = 'left')
````

````python
#right join
pd.merge(df6, df7, how = 'right')
````

````python
#inner join
pd.merge(df6, df7, how = 'inner')
````

````python
#outer join
pd.merge(df6, df7, how = 'outer')
````

# left_index, right_index

````python
df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})
df1.set_index('employee', inplace = True)
df1
````

````python
df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue', 'Mike'],
                    'hire_date': [2004, 2008, 2012, 2014, 2010]})
df2.set_index('employee', inplace = True)
df2
````

````python
df3 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'salary': [70000, 80000, 120000, 90000]})
df3
````

````python
pd.merge(df1, df2, left_index = True, right_index = True, how = 'left')
````

````python
pd.merge(df2, df3, left_index = True, right_on = 'name', how = 'left')
````

# left_on, right_on

````python
df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})
df1
````

````python
df3 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'salary': [70000, 80000, 120000, 90000]})
df3
````

````python
pd.merge(df1, df3, left_on = 'employee', right_on = 'name')
````

# on

````python
#If 2 DataFrames have a column with the same name
#and we want to merge these 2 base on that column
#then use pd.merge(df1, df2, on = 'column name')
df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})
````

````python
df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],
                    'hire_date': [2004, 2008, 2012, 2014]})
````

````python
on = pd.merge(df1, df2, on = 'employee')
on
````

# suffixes

````python
Math_score = pd.DataFrame([['Trung', 10], ['Kien', 9.5], ['Duc', 8.9]], columns = ['Name', 'Score'])
Math_score
````

````python
English_score = pd.DataFrame([['Trung', 10], ['Kien', 10], ['Duc', 9.5]], columns = ['Name', 'Score'])
English_score
````

````python
pd.merge(Math_score, English_score, on = 'Name', how = 'outer', suffixes = ['_math', '_english'])
````

````python
data = pd.merge(pop, abbr, left_on = 'state/region', right_on = 'abbreviation', how = 'outer').drop(columns = 'abbreviation')
data
````

````python
missing_data = data[data['state'].isna()]
missing_data
````

````python
missing_data['state/region'].unique()
````

````python
data.loc[data['state/region'] == 'PR', 'state'] = 'Peru'
data.loc[data['state/region'] == 'USA', 'state'] = 'United States'
data
````

````python
data.isna().any(axis = 0)
````

````python
full = pd.merge(data,areas, on = 'state', how = 'outer')
full
````

````python
#remove rows with missing values
full.dropna(axis = 'index', inplace = True)
full
````

````python
full['density'] = full['population'] / full['area (sq. mi)']
full
````

````python
full.set_index('state/region', inplace = True)
````

````python
full.sort_values(by = ['density', 'population'], ascending = False, kind = 'mergesort')
````
