```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./pandas/datasets/shampoo.csv')
title = df.columns[1]
print(df.columns)
rename = {}
rename[title] = 'Sales'
df.rename(columns = rename, inplace = True)
#c = Series.pct_change() => Series[i] = (1 + c[i]) * Series[i - 1] => the i-th element show how much the previous value changes
#compare to the current value
pct_change = df['Sales'].pct_change()

#c = Series.diff(periods = k) => c[i] = Series[i] - Series[i - k] => c[i] is the different between the i-th and (i - k)-th element
#by default periods = 1
diff = df['Sales'].diff()
fig, ax = plt.subplots(3,1,sharex = True)
ax[0] = plt.subplot(311)
ax[0].plot(range(len(df)), df['Sales'])
# xticks = np.arange(0, len(df), 2)
# ax.set_xticks(xticks)
# ax.set_xticklabels(df['Month'][xticks], rotation = 60, fontsize = 8, va = 'top')
ax[0].set_title(title)



ax[1].stem(range(len(pct_change) - 1), pct_change[1:], linefmt = '--', use_line_collection = True)
ax[1].set_ylim(-2, 2)

ax[2].stem(range(len(diff) - 1), diff[1:], linefmt = '--', markerfmt = 'd', use_line_collection = True)
plt.subplots_adjust(hspace = .4)
plt.show()
```
