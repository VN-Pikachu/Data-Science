
sort based on values of multiple columns:

````python
pokemon = pd.read_csv('./pokemon.csv')

#order by Attack, Defense, HP desc (SQL)
#is equivalent to: df.sort_values(by = ['Attack', 'Defense', 'HP'], ascending = False)
order = pokemon.sort_values(by = ['Attack', 'Defense', 'HP'], ascending = False, kind = 'heapsort')
order[['Name', 'Attack', 'Defense', 'HP', 'Legendary']]
````

````python
sort based on value of 1 column:
````

````python
#let's order pokemon by its speed descending
pokemon.sort_values(by = 'Speed', ascending = False, kind = 'mergesort', axis = 0)
````

````python
data = [
    [30, 'VN Champions', 56971],
    [35, 'Dirilis', 58930],
    [33, 'VN Champions', 57131]
]
TF = pd.DataFrame(data, index = ['VN Pikachu', 'ARMAGEDON', 'bac giang vn'], columns = ['level', 'clan', 'average damage']).T
TF
````

````python
#let's sort columns based on the value of row average damage desc inplace
TF.sort_values(axis = 1, by = 'average damage', ascending = False, kind = 'quicksort')
````
