`````python
electric = pd.read_csv('./electric_production.csv', index_col = 0)
electric.head()
`````

`````python
#7 days average
e_mean = electric.rolling(window = 7, min_periods = 1).mean()
e_mean.head()
`````

`````python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
electric.plot(ax = ax)
e_mean.plot(ax = ax, color = 'r')
`````

# expanding

```python
nums.expanding(min_periods=1, center=False, axis=0)
```

```python
min_periods : int, default 1
    Minimum number of observations in window required to have a value
    (otherwise result is NA).
center : bool, default False
    Set the labels at the center of the window.
axis : int or str, default 0
````

`````python
#cumsum
nums.expanding().sum()
`````

`````python
#start having value from index 5
nums.expanding(min_periods = 5).sum()
`````

`````python
matrix = pd.DataFrame(np.arange(12).reshape((3,4)), dtype = int)
matrix
`````

`````python
#cumstd each column
matrix.expanding(axis = 0).std()
`````

`````python
#cumsum each row
matrix.expanding(axis = 1).sum()
`````
