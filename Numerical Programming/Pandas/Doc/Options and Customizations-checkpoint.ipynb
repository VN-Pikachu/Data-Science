Pandas provide API to customize some aspects of its behavior, display is being mostly used.<br>

The API is composed of five relevant functions. They are:<br>

* get_option()
* set_option()
* reset_option()
* describe_option()
* option_context()

# get_option()

Available options:

- compute.[use_bottleneck, use_numexpr]
- display.[chop_threshold, colheader_justify, column_space, date_dayfirst,
  date_yearfirst, encoding, expand_frame_repr, float_format]
- display.html.[border, table_schema, use_mathjax]
- display.[large_repr]
- display.latex.[escape, longtable, multicolumn, multicolumn_format, multirow,
  repr]
- display.[max_categories, max_columns, max_colwidth, max_info_columns,
  max_info_rows, max_rows, max_seq_items, memory_usage, multi_sparse,
  notebook_repr_html, pprint_nest_depth, precision, show_dimensions]
- display.unicode.[ambiguous_as_wide, east_asian_width]
- display.[width]
- html.[border]
- io.excel.xls.[writer]
- io.excel.xlsm.[writer]
- io.excel.xlsx.[writer]
- io.hdf.[default_format, dropna_table]
- io.parquet.[engine]
- mode.[chained_assignment, sim_interactive, use_inf_as_na, use_inf_as_null]
- plotting.matplotlib.[register_converters]

# set_option()

```python

pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', 5)
```

```python
pokemon = pd.read_csv('./pokemon.csv')
pokemon
```

# reset_option()

```python
pd.reset_option('display.max_columns')
pd.get_option('display.max_columns')
```

```python
pokemon
```

# describe_option()

```python
pd.describe_option('display.width')
```

# option_context()

```python
with pd.option_context('display.max_rows', 20):
    print(pokemon)
```

# Frequently used

1	
<b>display.max_rows</b>

Displays maximum number of rows to display

2	
<b>display.max_columns</b>

Displays maximum number of columns to display

3	
<b>display.expand_frame_repr</b>

Displays DataFrames to Stretch Pages

4	
<b>display.max_colwidth</b>

Displays maximum column width

5	
<b>display.precision</b>

Displays precision for decimal numbers

```python
housing = pd.read_csv('./housing.csv')
housing
```

```python
with pd.option_context('display.precision', 2, 'display.max_columns', 6, 'display.width', 150):
    print(housing)
```

```python
pd.get_option('display.max_colwidth')
```

```python
with pd.option_context('display.max_colwidth', 10, 'display.max_columns', 5):
    print(pokemon)
```
