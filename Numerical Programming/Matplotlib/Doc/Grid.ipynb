````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
x = np.linspace(-2 * np.pi, 2 * np.pi, 50)
````

# Axes.grid

```python
Axes.grid(b=None, which='major', axis='both', **kwargs)
```

````python
fig = plt.figure(figsize = (10, 3))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

ax1.plot(x, np.sin(x))
ax2.plot(x, np.cos(x))
ax3.plot(x, np.tanh(x))

ax1.grid(linewidth = 1, linestyle = ':', color = 'dodgerblue')
ax2.grid(linewidth = 1, linestyle = '--', color = 'gold')
ax3.grid(False)

ax1.set_title('custom grid')
ax2.set_title('custom grid')
ax3.set_title('no grid')

plt.subplots_adjust(wspace = 0.5)
plt.show()
````
