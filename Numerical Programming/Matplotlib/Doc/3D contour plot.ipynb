```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
xv, yv = np.meshgrid(x,y)
zv = xv *xv + yv * yv
```

```python
ax = plt.subplot(projection = '3d')
ax.contour3D(xv, yv, zv, 30)
```
