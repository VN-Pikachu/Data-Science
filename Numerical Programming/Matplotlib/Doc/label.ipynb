````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
````

# Figure.align_label

````python
fig, axes = plt.subplots(1,2, figsize = (10,4), gridspec_kw = {'hspace': 0.2})
axes[0].plot(x,np.sin(x))
axes[1].plot(x,np.cos(x))
axes[0].set_xlabel('sine function')
axes[1].set_xticks(x)
axes[1].set_xticklabels(x, rotation = 30)
axes[1].set_xlabel('cos function')

fig.align_labels()
````
