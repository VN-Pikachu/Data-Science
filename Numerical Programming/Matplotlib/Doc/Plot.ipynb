````python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
````

````python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine Function')
plt.show()
````

````python
import numpy as np;
import matplotlib.pyplot as plt;

x = np.linspace(0,20,100)
y1 = np.sin(x)

y2 = np.sin(x) + np.cos(x)
ax1 = plt.subplot(211)
ax1.plot(x, y1, '-ob', mec = 'k', mew = '1', mfc = 'teal')

ax2 = plt.subplot(212, sharey = ax1)
ax2.plot(x, y2, '-or', mec = 'k', mfc = 'green', mew = .8)

plt.show()
````

# markevery

e.g. **`markevery=5`** will plot every 5th marker starting from the first data point.

````python
X = np.linspace(-10, 10, 100)
y = np.sin(X)
````

````python
#markevery = None (default): splot every marker
plt.plot(X, y, '-o')
````

````python
#markevery = 5, plot every 5-th marker starting from the first data point
plt.plot(X, y, '-o', markevery =5)
````

````python
#markevery = (8, 30): plot starting fro 8-th marker, step = 30 markers
plt.plot(X, y, '-o', markevery = (8, 30))
````

````python
#markevery = [16,24,30]: plot 16-th, 24-th, 30-th marker
plt.plot(X, y, '-o', markevery = [16,24,30])
````

````python
#markevery = [0,-1]: plot the first and last marker
plt.plot(X, y, '-o', markevery = [0,-1])
````

````python
fig, axes = plt.subplots(1,2)
axes[0].plot(X, y,'-o', markevery = 0.1)
axes[1].plot(X, y,'-o', markevery = .5)

axes[0].set_title('markevery = 0.1')
axes[1].set_title('markevery = 0.5')
fig.tight_layout()
````
