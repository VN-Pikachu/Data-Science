```python
import matplotlib.pyplot as plt;
import numpy as np
x = np.linspace(1,20,50)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sqrt(x) - 3

ax1 = plt.subplot(311)
ax1.plot(x,y1, '-b')

ax2 = plt.subplot(312)
ax2.plot(x, y2, '-b')
plt.setp(ax2.get_xticklabels(), visible = False) #turn off the xticklabels in the second subplot

#share the xlim of the third subplot to the second subplot
#share the ylim of the third subplot to the first subplot
ax3 = plt.subplot(313, sharex = ax2, sharey = ax1)
ax3.plot(x,y3, '-b')
ax3.set_ylim(-2,2)
ax3.set_xlim(0, 21)

plt.subplots_adjust(hspace = .5)
plt.show()
```
