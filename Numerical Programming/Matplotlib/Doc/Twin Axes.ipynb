# Axes.twinx()

return an Axes object share the same x axis

```python
fig = plt.figure(figsize = (10,10))
x = np.arange(2,10)
y1 = np.exp(x)
y2 = np.log(x)

ax = fig.add_subplot()
line = ax.plot(x, y1)
ax.set_ylabel('exp(x)')

ax1 = ax.twinx()
line_1 = ax1.plot(x, y2, 'r')
ax1.set_ylabel('log(x)')

fig.legend(labels = ('exp(x)', 'log(x)'), loc = 'upper left')
```

```python
import matplotlib.pyplot as plt;
import numpy as np;
from matplotlib import style
style.use('dark_background')

#use ax2 = ax1.twinx() to plot in the same subplot, share the x-axis
months = np.arange(12)
labels = ['I', 'II', 'III', 'IV', 'V', 'VI', 'VII', 'VIII', 'IX', 'X', 'XI', 'XII']
rain = np.random.randint(10, 250, 12)
temp = np.random.rand(12) * 6 + 22
fig = plt.figure()
ax1 = fig.gca()
ax1.bar(labels, rain, color = 'white')
ax1.set_ylim(0,350)
ax1.tick_params(color = 'white', labelcolor = 'dodgerblue')
ax1.set_xlabel('Months', color = 'aqua')
ax1.set_ylabel('Rain')
ax1.set_title('Rain and Temperature in Ha Noi 2018')
ax2 = ax1.twinx()
ax2.plot(months, temp, '-or', ms = 4, mfc = 'white', mec = 'aqua')
ax2.set_ylim(0, 36)
ax2.set_yticks(np.arange(0,36,2))
ax2.tick_params(axis = 'y', color = 'yellow', labelsize = 8, labelcolor = 'orange')
ax2.set_ylabel('Temperature', rotation = -90, va = 'bottom')
plt.show()
```
