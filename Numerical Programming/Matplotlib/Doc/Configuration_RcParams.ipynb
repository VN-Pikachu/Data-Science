```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

**`rcParams`** is a **dict** 

```python
rcParams
```

```python
x = np.linspace(-10, 10, 100)
y = np.sinc(x)
```

Modify settings by modifying the values of **`rcParams`** 

```python
rcParams['grid.color'] = 'gold'
rcParams['grid.linestyle'] = '--'
rcParams['grid.linewidth'] = 5
plt.plot(x, y)
plt.grid(True)
```

using **`plt.rc`** for more convenience

```python
#instead of using
#rcParams['grid.linewidth'] = 1
#rcParams['grid.linestyle'] = ':'
#rcParams['grid.color'] = 'red'
plt.rc('grid', linewidth = 1, linestyle = ':', color = 'red')
plt.plot(x, y)
plt.grid(True)
```

```python
plt.rc('axes', facecolor = 'lightgray', grid = True) 
plt.rc('grid', color = 'w', linewidth = 1, linestyle = '-')
plt.rc('lines', linestyle = '-', color = 'r', linewidth = 5)
plt.rc('xtick', color = 'teal')
plt.rc('ytick', color = 'teal') #equivalent: plt.rcParams['ytick.color'] = 'teal'
x = np.linspace(0,15, 100)
y = np.sin(x)
plt.plot(x, y)
```
