````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import patches
````

````python
x = np.linspace(0, 2 * np.pi, 50)
y = np.sin(x)
````

# Adding lines

Create a Line2D Object: <code><b>plt.Line2D</b></code>

```python
plt.Line2D(
    xdata,
    ydata,
    linewidth=None,
    linestyle=None,
    color=None,
    marker=None,
    markersize=None,
    markeredgewidth=None,
    markeredgecolor=None,
    markerfacecolor=None,
    markerfacecoloralt='none',
    fillstyle=None,
    antialiased=None,
    dash_capstyle=None,
    solid_capstyle=None,
    dash_joinstyle=None,
    solid_joinstyle=None,
    pickradius=5,
    drawstyle=None,
    markevery=None,
    **kwargs,
)
```

Adding Line2D object to Axes: <code><b>Axes.add_line(Line2D_object)</b></code>

````python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
for p, q in zip(x, y):
    ax.add_line(plt.Line2D((p,0), (0,q), color = 'teal'))
````

# Adding shapes

````python
help(patches)
````

````python
fig, axes = plt.subplots(1,3,subplot_kw = dict(xticks = [], yticks = []))
#Adding a circle
circle = patches.Circle(xy = (0,0), radius = 3)
axes[0].add_patch(circle)
axes[0].set_aspect('equal')
axes[0].set(xlim = (-5,5), ylim = (-5,5))

#Adding a Rectangle
rect = patches.Rectangle(xy = (0,0), width = 10, height = 5, color = 'gold')
axes[1].add_patch(rect)
axes[1].set(xlim = (-10, 20), ylim = (-10, 20))

#Adding Ellipse
elip = patches.Ellipse(xy = (0,0), width = 5, height = 2, angle = 45)
axes[2].add_patch(elip)
axes[2].set(xlim = (-5,5), ylim = (-5,5))
````

````python
#adding polygon
x = np.linspace(0, 2 * np.pi, 7)
points = np.vstack((np.cos(x), np.sin(x))).T
points
````

````python
fig = plt.figure()
ax = fig.add_subplot()
ax.add_patch(patches.Polygon(points))
ax.set(xlim = (-2,2), ylim = (-2,2))
ax.grid(True)
````

````python
#adding Fancybox
fancybox = patches.FancyBboxPatch(xy = (0,0), width = 10, height = 3, boxstyle='darrow')
plt.gca().add_patch(fancybox)
plt.xlim((-5,15))
plt.ylim((-5,10))
````
