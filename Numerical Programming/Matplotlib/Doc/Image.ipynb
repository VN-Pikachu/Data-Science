# Origin

Place the [0,0] index of the array in the upper left or lower left corner of the axes

```python
#orign = {'lower', 'upper'}
fig, axes = plt.subplots(1,2, figsize = (20,20), dpi = 500)
axes[0].imshow(img, origin = 'upper')
axes[0].set_xlabel('origin = upper')

axes[1].imshow(img, origin = 'lower')
axes[1].set_xlabel('origin = lower')
```

# Interpolation

```python
#let's visualize the result of some interpolation methods
methods = ['nearest', 'bilinear', 'bicubic']
X = np.random.rand(5,5)
fig, axes = plt.subplots(1,3, subplot_kw = dict(xticks=[], yticks =[]))
for ax, interp in zip(axes.flat, methods):
    ax.imshow(X, interpolation = interp)
    ax.set_title(interp.capitalize())
```

# cmap

```python
simple_img = img[:,:,0] #1 channel
plt.imshow(simple_img) 
```

```python
plt.imshow(simple_img, cmap = 'hot')
```

Note that you can also change colormaps on existing plot objects using the **`set_cmap()`** method:

```python
imgplot = plt.imshow(simple_img)
imgplot.set_cmap('hot')
```

# extent

**`(left, right, bottom, top)`**

```python
fig = plt.figure(figsize = (15,10), facecolor = 'teal')
ax = fig.add_subplot(fc = 'red')
ax.imshow(img, extent = (0, 2,0, 1))
#ax.imshow(img)
```
