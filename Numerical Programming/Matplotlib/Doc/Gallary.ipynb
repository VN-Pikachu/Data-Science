```python
import numpy as np 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
labels = ['I', 'II', 'III', 'IV', 'V', 'VI', 'VII', 'VIII', 'IX', 'X', 'XI', 'XII']
rain = np.random.randint(10, 250, 12)
temp = np.random.rand(12) * 6 + 22
```

```python
fig = plt.figure(facecolor = 'k')
ax = fig.add_axes([0,0,1,1], fc = 'k')
months = np.arange(12)
ax.bar(labels, rain, color = 'white', label = 'rain')
ax.set_ylim(0, 350)
ax.set_xlabel('Months', color = 'white')
ax.set_ylabel('Rain', color = 'white')
ax.set_title('Rain and Temperature', color = 'gold')

for pos in ax.spines:
    ax.spines[pos].set(color = 'white', linewidth = 2)
ax.tick_params(axis = 'x', color = 'red', direction = 'out', labelcolor = 'teal', labelsize = 12, width = 5)
ax.tick_params(axis = 'y', color = 'blue', labelcolor = 'blue', width = 2, length = 10)
ax1 = ax.twinx()
ax1.plot(months, temp, '-or', mfc = 'w', mec = 'teal', label = 'temperature')
ax1.set(ylabel = 'temperature', ylim = (0,34))
ax1.tick_params(axis = 'y', labelcolor = 'yellow', color = 'pink', width = 2, length = 5, labelsize = 8)
ax1.set_ylabel('Temperature', color = 'red', rotation = -90, va = 'bottom')

fig.legend(bbox_to_anchor = (0, 1, 1, 0), loc = 'upper left', ncol = 2, mode = 'expand', bbox_transform = ax.transAxes, title = 'Measurements', edgecolor = 'red')
```

```python
import numpy as np
import matplotlib.pyplot as plt;
#plt.rc('figure', autolayout = True)
fig = plt.figure(figsize = (12, 6))
category_names = ['Strongly disagree', 'Disagree',
                  'Neither agree nor disagree', 'Agree', 'Strongly agree']
results = {
    'Question 1': [10, 15, 17, 32, 26],
    'Question 2': [26, 22, 29, 10, 13],
    'Question 3': [35, 37, 7, 2, 19],
    'Question 4': [32, 11, 9, 15, 33],
    'Question 5': [21, 29, 5, 5, 40],
    'Question 6': [8, 19, 5, 30, 38]
}
groups = list(results.keys())
data = np.array(list(results.values()))

m, n = data.shape
color = plt.get_cmap('RdYlGn')(np.linspace(.15,.85,n))
cum = np.cumsum(data, axis = 1)
for i in range(n):
    left = cum[:,i-1] if i else np.zeros(m)
    rects = plt.barh(groups, data[:,i], left = left, color = color[i], alpha = .8, height = .75, label = category_names[i])
    for j ,rect in enumerate(rects):
        y, h = rect.get_y(), rect.get_height()
        #red, green, blue= color
        text_color = 'darkgray' if np.product(color[i]) > .5 else 'white'
        plt.text(left[j] + data[j,i] / 2, y + h /2, data[j,i], ha = 'center', va = 'center', color = text_color)
ax = plt.gca()
ax.invert_yaxis()
ax.xaxis.set_visible(False)
ax.legend(bbox_to_anchor = (0,1,1,1), loc = 'lower left', mode = 'expand', borderaxespad = .2, ncol = n)
ax.set_xlim(0, cum[:,-1].max())
plt.show()
```
