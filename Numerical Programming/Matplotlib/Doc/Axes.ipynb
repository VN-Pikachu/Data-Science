````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
x = np.linspace(0, 2 * np.pi, 50)
y = np.cos(x)
````

# Figure.add_axes([left, bottom, width, height])

```python
add_axes(rect, projection=None, polar=False, **kwargs)
add_axes(ax)
```

Adds an axes at position rect [left, bottom, width, height] where all quantities are in fractions of figure width and height.

````python
fig = plt.figure(figsize = (5,5), facecolor = 'lime', linewidth = 3, edgecolor = 'gold', dpi = 200)

ax1 = fig.add_axes([0, 0.25, 0.4, 0.4])

ax2 = fig.add_axes([0.5, 0.25, 0.4, 0.4])

ax1.plot(x, np.sin(x), '-sr', ms = 1, mec = 'gold')
ax2.plot(x, np.cos(x), linestyle = '-', marker = 's', color = 'r', ms = 1, mec = 'gold')
plt.show()
````

# scaling both axes equally

<code><b>Axes.set_aspect('equal')</b></code>

````python
fig = plt.figure()
ax = fig.add_subplot()

ax.add_patch(plt.Circle(xy = (0,0), radius = 1, color = 'gold', ls = ':', lw = 2, ec = 'red'))
ax.set(xlim=(-2,2), ylim=(-2,2))
#without equal aspect, circle looks like ellipse
#let's fix this 
ax.set_aspect('equal')
````

# Log Scale

<code><b>Axes.xscale()</b></code><br>
<code><b>Axes.yscale()</b></code>

````python
fig = plt.figure()
ax = fig.add_subplot()
X = np.logspace(1, 2, 100)
y = np.log(X)

ax.plot(X, y)
#ax.set_xscale('symlog')
````

# Polar System

````python
fig = plt.figure()
ax = fig.add_subplot(polar = True)

X = np.linspace(0, 2 * np.pi, 50)
y = np.sin(X)

ax.plot(X, y)
````
