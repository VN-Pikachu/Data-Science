## bbox

```python
import numpy as np
import matplotlib.pyplot as plt;

np.random.seed(101)
cov = [[1, 0],[0,1]]
N = 50
ax = plt.subplot(211)
data1 = np.random.multivariate_normal([1, 1], cov, N)
data2 = np.random.multivariate_normal([5,5], cov, N)
ax.scatter(data1[:,0], data1[:,1], c = 'green')
ax.scatter(data2[:,0], data2[:,1], c = 'red')
ax.text(2, 3, 'direction', bbox = dict(boxstyle = 'rarrow,pad=1', fc = 'cyan', ec = 'blue'), rotation = 45, ha='center', va='center')
ax.text(0, 0, 'sample A', bbox = dict(boxstyle = 'square, pad = 1', fc = 'white'), ha = 'center', va = 'center')
ax.text(6, 6, 'sample B', bbox = dict(boxstyle = 'round, pad = 1', fc = 'white'), ha = 'center', va = 'center')

ax = plt.subplot(212)
styles = ['circle', 'darrow', 'rarrow', 'larrow', 'square', 'round', 'round4', 'roundtooth', 'sawtooth']
ax.set_xticks(())
ax.set_yticks(())
for i, style in enumerate(styles):
    ax.text(i // 3, (i % 3) * 3, style, bbox = dict(boxstyle = '{},pad=.5'.format(style)), ha = 'center', va = 'center')
ax.set_ylim(-2, 8)
ax.set_xlim(-1, 3)
plt.show()
```

```python
boxstyle = 'circle'

fig = plt.figure(figsize = (5,5))
ax = fig.add_axes([0,0,1,1])
ax.set(xlim = (0,2), ylim = (0,2))
ax.text(1,1, 'VN Pikachu', bbox = dict(boxstyle = f'{boxstyle},pad=1', fc = 'teal', ec = 'red'), rotation = 30, ha = 'center', va = 'center')
```

# annotate

```python
help(plt.annotate)
```

```python
ax = plt.subplot()

ax.annotate('hello', arrowprops = {'arrowstyle': '->'}, xy = (2,1), xytext = (3,4),  ha = 'center', va = 'center')
ax.set_xlim(0, 5)
ax.set_ylim(0,5)
```

## arrowprops

<table>
    <tr><th>arrowprops key</th><th>	description</th></tr>
    <tr><td>width</td><td>	the width of the arrow in points</td></tr>
    <tr><td>frac</td><td>	the fraction of the arrow length occupied by the head</td></tr>
    <tr><td>headwidth</td><td>	the width of the base of the arrow head in points</td></tr>
    <tr><td>shrink</td><td>	move the tip and base some percent away from the annotated point and text</td></tr>
    <tr><td>**kwargs</td><td>	any key for <code>matplotlib.patches.Polygon</code>, e.g., <i>facecolor</i></td></tr>

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
props = {
    'arrowstyle': '->',
    'connectionstyle' : "angle3,angleA=90,angleB=0"
}
style = '->'
ax.annotate('hello', arrowprops = props, xy = (2,1), xytext = (3,4),  ha = 'center', va = 'center')
ax.set(xlim = (0,5), ylim = (0,5))
```
