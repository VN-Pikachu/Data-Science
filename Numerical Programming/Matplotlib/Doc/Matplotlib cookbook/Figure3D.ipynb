# Plot a 3D parametric surface

```python
r = 4
d = 10
phi = np.linspace(0, 2 * np.pi, 36)
theta = np.linspace(0, 2 * np.pi, 36)
phi, theta = np.meshgrid(phi, theta)
z = r * np.sin(theta)
x = (d + r * np.cos(theta)) * np.cos(phi)
y = (d + r * np.cos(theta)) * np.sin(phi)
ax = plt.subplot(projection = '3d')
ax.plot_surface(x, y, z)
ax.set_zlim(-10, 10)
```

```python
r = 4
d = 15
phi = np.linspace(0, 2 * np.pi, 36)
theta = np.linspace(0, 2 * np.pi, 36)
phi, theta = np.meshgrid(phi, theta)
z = r * np.sin(theta)
x = (d + r * np.cos(theta)) * np.cos(phi)
y = (d + r * np.cos(theta)) * np.sin(phi)
ax = plt.subplot(projection = '3d')
ax.plot_wireframe(x, y, z, color = 'teal')
ax.set_zlim(-10, 10)
```

# Embedding 2D figures in a 3D figure

```python
X = np.linspace(-3, 3)
Y = np.linspace(-3, 3)
x, y = np.meshgrid(X, Y)
u = np.exp(- X * X)
z = np.exp(-(x * x + y * y))
ax = plt.subplot(projection = '3d')
ax.plot_surface(x, y, z, cmap = 'Spectral')
ax.plot(X, u, zdir = 'x', zs = -3) #plot on a plane that the x-axis is orthonomal to and crosses the x-axis at x = -3
ax.plot(X, u, zdir = 'y', zs = 3)
ax.set_zlim(0, 3)
```
