# Adding lines

## How to add a single line?

<p class = 'note'>
    Create a line object: <code>line = plt.Line2D((x1, x2), (y1, y2), **kwgs)</code><br>
    Add that line to the current axes: <code>ax.add_line(line)</code>
</p>

```python
ax = plt.gca()
line1 = plt.Line2D((0, 5), (0, 5), ls = '--', linewidth = 5, color = 'red')
line2 = plt.Line2D((0, 5), (5, 0), ls = '--', linewidth = 5, color = 'g')
ax.add_line(line1)
ax.add_line(line2)
plt.axis('scaled')
```

```python
N = 16
for i in range(N):
    plt.gca().add_line(plt.Line2D((0, i), (N - i , 0), c = 'aqua'))
plt.axis('scaled')
```

```python
np.random.seed(6)
X = np.linspace(0, 20, 20)
y = X + 3 * np.random.randn(len(X))
plt.scatter(X, y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X[:, None], y)
```

```python
y_pred = model.predict(X[:,None])
plt.scatter(X, y)
plt.plot(X, y_pred, lw = 3, c = 'r')
#plot residual
for x, y1, y2 in zip(X, y, y_pred):
    line = plt.Line2D((x, x), (y1, y2), ls = '--', color = 'g')
    plt.gca().add_line(line)
#Parameters
plt.title('Slope: %.2f \n Intercept: %.f' % (model.coef_[0], model.intercept_))
```

# Adding shapes

<p class = 'note'>
    Import <code>patches</code>: <code>from matplotlib import patches</code><br>
    Create a shape object: <code>shape = patches.Circle(**kwgs)</code> 
    <span class = 'tag'>plt.Polygon, plt.Rectangle, ...</span>
    <br>
    attach that shape to the axes: <code>ax.add_patch(shape)</code>


```python
from matplotlib import patches
```

```python
# plt.circle?
shape = patches.Circle((0, 0), 5, fc = 'dodgerblue', alpha = .3)
plt.gca().add_patch(shape)
plt.axis('scaled')
```

```python
#plt.Rectangle?
shape = patches.Rectangle((0, 0), 10, 5, hatch = 'O')
plt.gca().add_patch(shape)
plt.axis('scaled')
```

```python
shape = patches.Ellipse((0, 0), 5, 2, fc = 'g', alpha = .2, ls = ':')
plt.gca().add_patch(shape)
plt.axis('scaled')
```

```python
N = 9
angles = np.linspace(0, 2 * np.pi, N)
points = np.c_[np.cos(angles), np.sin(angles)]
shape = patches.Polygon(points, ls = '-.', fc = 'none', lw = 3, color = 'teal')
plt.gca().add_patch(shape)
plt.axis('scaled')
```
