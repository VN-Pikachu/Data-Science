# Controlling the number of ticks

```python
X = np.linspace(-10, 10, 200)
plt.plot(X, np.sin(X), label = 'sine', color = 'green')
plt.plot(X, np.cos(X), label = 'cos', color = 'red')
```

<p class = 'note'>Now let's try to set the number of major ticks on the x axis and the y axis to be 5</p>

```python
X = np.linspace(-10, 10, 200)
plt.plot(X, np.sin(X), label = 'sine', color = 'green')
plt.plot(X, np.cos(X), label = 'cos', color = 'red')

ax = plt.gca()
ax.xaxis.set_major_locator(plt.MaxNLocator(5))
ax.yaxis.set_major_locator(plt.MaxNLocator(5))
```

# Changing properties of Labels

<p class = 'note'>Use: <code>ax.tick_params(**kwgs)</code></p>

```python
import pandas as pd
print(os.getcwd())
ax = plt.subplot()
df = pd.read_csv('Electric_production.csv', index_col = 0, parse_dates = True)
df.plot(ax = ax)
df.rolling(7).mean().plot(color = 'red', ax = ax)
ax.tick_params(axis = 'x', rotation = 45)
```
