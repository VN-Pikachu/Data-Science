```python
x = np.linspace(-np.pi,np.pi, 50)
y = np.sin(x)
```

```python
style.use('seaborn-ticks')
x = np.linspace(0, 2 * np.pi, 100)

fig, ax = plt.subplots()
ax.plot(x, np.sin(x), label = 'sin')
ax.plot(x, np.cos(x), label = 'cos')

ax.legend(bbox_to_anchor = (0,1,1,1), mode = 'expand', loc = 'lower left', ncol = 2, frameon = True, 
          borderaxespad = 0, title = 'Functions', fancybox = True, shadow = True,
          facecolor = 'dodgerblue', framealpha = .1, edgecolor = 'r')
ax.tick_params(top = False, right = False)

ax.xaxis.set_major_locator(plt.MultipleLocator(.5))
ax.yaxis.set_minor_locator(plt.MultipleLocator(.1))
ax.yaxis.set_major_locator(plt.MaxNLocator(3))
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda value, position: f'{int(value)}$'))
ax.grid(True, which = 'both')

ax.tick_params(axis = 'x', labelcolor = 'gold', labelrotation = 30)
```
