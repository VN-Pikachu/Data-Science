```python
import numpy as np 
import matplotlib.pyplot as plt
```

# plt.axhline(y=0, xmin=0, xmax=1, **kwargs)

draw a horizontal line at y from xmin to xmax<br>
---------<br>
y : scalar, optional, default: 0
    y position in data coordinates of the horizontal line.

xmin : scalar, optional, default: 0
    Should be between 0 and 1, 0 being the far left of the plot, 1 the
    far right of the plot.

xmax : scalar, optional, default: 1
    Should be between 0 and 1, 0 being the far left of the plot, 1 the
    far right of the plot.


<b style = 'color:red'>NOTE: <code>xmin</code> and <code>xmax</code> is in Axes scale(between 0 and 1)</b>

# plt.axvline(x=0, ymin=0, ymax=1, **kwargs)

draw a vertical line at x from ymin to ymax

```python
x = np.linspace(0, 2 * np.pi, 15)
y = np.sin(x)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x, y, '-o')
ax.set_title('Sine Function')

ax.axhline(y = 1, xmin = 0, xmax = 1, color = 'r', ls = '--')
ax.axhline(y = -1, xmin = 0, xmax = 1, color = 'r', ls = '--')
ax.axvline(x = 0, ymin = 0.25, ymax = 0.75, color = 'y')
ax.axvline(x = 2 * np.pi, ymin = 0.25, ymax = 0.75, color = 'y')

ax.annotate('axhline', xy = (1, -1), xytext = (1, -0.75), arrowprops = dict(arrowstyle = 'fancy'), transform = ax.transData)
ax.annotate('axvline', xy = (0, -.25), xytext = (1, -.25), arrowprops = dict(arrowstyle = '-|>'), transform = ax.transData)
```
