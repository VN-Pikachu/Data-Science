````python
import numpy as np 
import matplotlib.pyplot as plt
````

# How to set position and labels for ticks

````python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
X = np.linspace(0, 2 * np.pi, 20)
y = np.sin(X)

ax.plot(X, y)

ax.set_xticks([0, np.pi / 2, np.pi])
ax.set_xticklabels(['0', 'pi/2', 'pi'])

ax.set_yticks([-1, -.5, 0, .5, 1])
ax.set_yticklabels(['-1', '-1/2', '0', '1/2', '1'])
````

# How to turn off xaxis and yaxis

**`Axes.xaxis.set_visible(False)`**  
**`Axes.yaxis.set_visible(False)`**

````python
x = np.linspace(-10, 10, 1000)
y = np.sinc(x)
fig, ax = plt.subplots()
ax.plot(x,y)
ax.xaxis.set_visible(False) #turn off x-axis (major ticks, minor ticks, ticklabels, xlabel)
ax.yaxis.set_visible(False) #turn off y-axis (major ticks, minor ticks, ticklabels, ylabel)
````

# How to remove xticklabels, yticklabels

set <code>labelbottom</code>, <code>labeltop</code>, <code>labelleft</code>, <code>labelright</code> to False

```python

Axes.tick_params(axis = 'x', labelbottom = False) #turn off labelbottom

Axes.tick_params(axis = 'y', labelleft = False) #turn off labelleft
```

# Axes.tick_params(**kwargs)

Signature: ax.tick_params(axis='both', **kwargs)
Docstring:
Change the appearance of ticks, tick labels, and gridlines.

Parameters
----------
axis : {'x', 'y', 'both'}, optional
    Which axis to apply the parameters to.

Other Parameters
----------------

axis : {'x', 'y', 'both'}
    Axis on which to operate; default is 'both'.

reset : bool
    If *True*, set all parameters to defaults
    before processing other keyword arguments.  Default is
    *False*.

which : {'major', 'minor', 'both'}
    Default is 'major'; apply arguments to *which* ticks.

direction : {'in', 'out', 'inout'}
    Puts ticks inside the axes, outside the axes, or both.

length : float
    Tick length in points.

width : float
    Tick width in points.

color : color
    Tick color; accepts any mpl color spec.

pad : float
    Distance in points between tick and label.

labelsize : float or str
    Tick label font size in points or as a string (e.g., 'large').

labelcolor : color
    Tick label color; mpl color spec.

colors : color
    Changes the tick color and the label color to the same value:
    mpl color spec.

zorder : float
    Tick and label zorder.

bottom, top, left, right : bool
    Whether to draw the respective ticks.

labelbottom, labeltop, labelleft, labelright : bool
    Whether to draw the respective tick labels.

labelrotation : float
    Tick label rotation

grid_color : color
    Changes the gridline color to the given mpl color spec.

grid_alpha : float
    Transparency of gridlines: 0 (transparent) to 1 (opaque).

grid_linewidth : float
    Width of gridlines in points.

grid_linestyle : string
    Any valid `~matplotlib.lines.Line2D` line style spec.

````python
x_params = {
    'axis': 'x',
    'direction':'in',
    'length': 10,
    'width': 5,
    'color': 'gold',
    'pad': 10,
    'labelsize': 20,
    'labelcolor': 'teal',
    'labelrotation': -30,
    'grid_color': 'green', 
    'grid_linestyle': '--',
    'grid_linewidth': 1,
}
````

````python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
ax.plot(x, y)
ax.grid(True)
ax.tick_params(**x_params)
#turn off ylabel
ax.tick_params(axis = 'y', labelleft = False)
````

# Controlling tick spacing

````python
fig = plt.figure()
ax = fig.add_subplot()
X = np.linspace(-15,15, 1024)
y = np.sinc(X)
ax.plot(X, y)

ax.xaxis.set_major_locator(plt.MultipleLocator(5)) #major ticks: step: 5 units
ax.xaxis.set_minor_locator(plt.MultipleLocator(1)) #minor ticks: step: 1 units

ax.grid(True, ls = '--', which = 'both') #set grid on both minor ticks and major ticks
````

# Controlling maximum number of ticks

````python
fig = plt.figure()
ax = fig.add_subplot()
X = np.linspace(-15,15, 1024)
y = np.sinc(X)
ax.plot(X, y)

ax.xaxis.set_major_locator(plt.MaxNLocator(3)) #3 major ticks on the x-axis
ax.yaxis.set_major_locator(plt.MaxNLocator(4)) #4 major ticks on the y-axis
````

# Format ticks

````python
fig = plt.figure()
ax = fig.add_subplot()
X = np.linspace(-15,15, 1024)
y = np.sinc(X)
ax.plot(X, y)

ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda value, position: f'{value}$'))
````

# OOP: Ticks and labels

Get ticklabel objects:

* **`Axes.xaxis.get_ticklabels()`**
* **`Axes.yaxis.get_tickalbels()`**


````python
X = np.linspace(-10, 10, 100)
y = np.sinc(X)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(X, y)

ticklabels = ax.xaxis.get_ticklabels() #a list of ticklabel object
#let rotate label by 45degree
for ticklabel in ticklabels:
    ticklabel.set(rotation = 45, color = 'green')
````

# Locators

````python
import numpy as np
import matplotlib.pyplot as plt


def tickline():
    plt.xlim(0, 10), plt.ylim(-1, 1), plt.yticks([])
    ax = plt.gca()
    ax.spines['right'].set_color('none')
    ax.spines['left'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['bottom'].set_position(('data',0))
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_minor_locator(plt.MultipleLocator(0.1))
    ax.plot(np.arange(11), np.zeros(11))
    return ax

locators = [
                'plt.NullLocator()',
                'plt.MultipleLocator(1.0)',
                'plt.FixedLocator([0, 2, 8, 9, 10])',
                'plt.IndexLocator(3, 1)',
                'plt.LinearLocator(5)',
                'plt.LogLocator(2, [1.0])',
                'plt.AutoLocator()',
            ]

n_locators = len(locators)

size = 512, 40 * n_locators
dpi = 72.0
figsize = size[0] / float(dpi), size[1] / float(dpi)
fig = plt.figure(figsize=figsize, dpi=dpi)
fig.patch.set_alpha(0)


for i, locator in enumerate(locators):
    plt.subplot(n_locators, 1, i + 1)
    ax = tickline()
    ax.xaxis.set_major_locator(eval(locator))
    plt.text(5, 0.3, locator[3:], ha='center')

plt.subplots_adjust(bottom=.01, top=.99, left=.01, right=.99)
plt.show()
````
