```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
xv, yv = np.meshgrid(x,y)
zv = xv *xv + yv * yv
```

# <code>Axes.plot_surface</code>

```python
ax = plt.subplot(projection = '3d')
ax.plot_surface(xv, yv, zv)
```

```python
alpha, beta = np.mgrid[0:2*np.pi:100j, 0:2*np.pi:100j]
R, D = 10, 2
x = (R + D * np.cos(beta)) * np.cos(alpha)
y = (R + D * np.cos(beta)) * np.sin(alpha)
z = D * np.sin(beta)
```

```python
fig = plt.figure(figsize = (8,4))
ax = fig.add_subplot(projection = '3d')
ax.plot_surface(x, y, z, lw = 3, cmap = 'twilight_r', rstride = 1, cstride = 1)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride = 1, cstride = 1, cmap = 'hot')
ax.set(xlim = (-4,4), ylim = (-4,4), zlim = (-2,2))
ax.contourf(X, Y, Z, offset = -2, cmap = 'hot')
```
