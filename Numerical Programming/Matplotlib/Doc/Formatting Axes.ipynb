```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

# Label

```python
iris = load_iris()
data, target = iris.data, iris.target
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.scatter(data[:,0], data[:,1], c = target, cmap = 'BrBG')
ax.set_xlabel(iris.feature_names[0])
ax.set_ylabel(iris.feature_names[1])
ax.set_title('Features correlation')
plt.show()
```

<h1> Axes.set_xticks<br>
Axes.set_yticks<br>
Axes.set_xticklabels<br> Axes.set_yticklabels
</h1>\

```python
x = np.linspace(0,np.pi, 100)
y = np.sin(x)
```

```python
fig = plt.figure(figsize = (6,6))
ax = fig.add_subplot()
ax.plot(x, y)

xticks = np.array([0, 1, 3 / 2, 2, 3]) * np.pi / 3
ax.set_xticks(xticks)
ax.set_xticklabels(['0', 'pi/3', 'pi/2', '2pi/3','pi'])

yticks = [0, np.sqrt(3) / 2, 1]
ax.set_yticks(yticks)
ax.set_yticklabels(['0', 'sqrt(3)/2', '1'])
plt.show()
```

# Scale

`Axes.set_xscale`<br>
`Axes.set_yscale`

```python
fig = plt.figure()
ax = fig.add_subplot()
x = np.linspace(0, 10, 100)
y = np.exp(x)

ax.plot(x, y)
#ax.set_xscale('log')
plt.show()
```

# Limit

`ax.set_xlim(xmin, xmax)`<br>
`ax.set_ylim(ymin, ymax)`

```python
x = np.linspace(0, 2 * np.pi, 50)
y = np.sin(x)

fig, ax = plt.subplots(1,2, figsize = (8, 4), gridspec_kw = {'wspace' : 1/3})
fig.suptitle('Sine Function')
ax[0].plot(x, y)
ax[1].plot(x, y)

ax[1].set_xlim(np.pi, 2 * np.pi)
plt.show()
```
