# bbox_to_anchor

```python
import numpy as np
import matplotlib.pyplot as plt;

X = np.linspace(0,2 * np.pi, 50)
y1 = np.sin(X)
y2 = np.cos(X)

plt.subplot(311)
plt.plot(X,y1, label = 'Sine')
plt.plot(X, y2, label = 'Consine')
#(x, y, width, height)
plt.legend(bbox_to_anchor = (0,1,1,1), ncol = 2, borderaxespad = .2, loc = 'lower left')
plt.xticks(())
plt.yticks(())

plt.subplot(312)
plt.plot(X,y1, label = 'Sine')
plt.plot(X, y2, label = 'Consine')
#(x, y, width, height)
plt.legend(bbox_to_anchor = (0,1,1,1), ncol = 2, borderaxespad = .2, loc = 'lower right')
plt.xticks(())
plt.yticks(())

plt.subplot(313)
plt.plot(X,y1, label = 'Sine')
plt.plot(X, y2, label = 'Consine')
#(x, y, width, height)
#Note: bbox_transform = ax.transData --> width, height is pixel
#      bbox_transform = ax.transAxes --> width, height is ratio of the axes
#      bbox_transform = fig.transFigure --> width, heigth is ratio of the figure
#if you set loc = 'lower right', it's the same effect
plt.legend(bbox_to_anchor = (0,1,1,1), ncol = 2, borderaxespad = .2, loc = 'lower left', mode = 'expand')
plt.xticks(())
plt.yticks(())

plt.subplots_adjust(hspace = .5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt;
from itertools import combinations

fig, axes = plt.subplots(4,4,subplot_kw = {'xticks':[], 'yticks':[]}, gridspec_kw = {'hspace': 1, 'wspace':.1}, figsize = (10,10))


xy = [(0, 0), (1,0), (0,1), (0, 1)]
locs = ['upper left', 'lower left', 'upper right', 'lower right']
X = []
y1 = []
y2 = []
for i in range(4):
    for j in range(4):
        ax = axes[i,j]
        ax.plot(X, y1,color = 'green', label = 'Sine')
        ax.plot(X, y2, color = 'red', label = 'Consine')
        ax.legend(bbox_to_anchor = xy[i], loc = locs[j], framealpha = .1, facecolor = 'cyan', edgecolor = 'gold', fontsize = 5, ncol = 2)
        if i == 0:
            ax.set_title(locs[j], fontweight = 'bold')
        if j == 0:
            ax.set_ylabel(str(xy[i]), fontweight = 'bold')
plt.show()
```

# Multiple legends

add `legend` (artirst) to `Axes`

```python
x = np.linspace(0, 2 * np.pi, 100)
```

```python
fig, ax = plt.subplots()
line1, line2, line3 = ax.plot(x, np.sin(x), '-r', x, np.cos(x), '--g', x, np.tanh(x), ':b')

legend1 = plt.legend(handles = [line1, line2], labels = ['Sine', 'Consine'], loc = 'lower left')
legend2 = plt.legend(handles = [line3], labels = ['Tanh'], loc = 'lower right')

ax.add_artist(legend1)
ax.add_artist(legend2)
```
