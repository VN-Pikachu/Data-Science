```python
import matplotlib.pyplot as plt;
import numpy as np
from matplotlib import style
np.random.seed(101)
#to use a comtempary style use: style.context('style-name')
def bar(code):
    x = np.random.randn(1000)
    plt.subplot(code)
    plt.hist(x, bins = 20)


```

```python
with style.context('dark_background'): bar(311)
```

```python
with style.context('ggplot'): bar(311)
```

```python
with style.context('bmh'): bar(311)
```

```python
with style.context('fivethirtyeight'): bar(311)
```
