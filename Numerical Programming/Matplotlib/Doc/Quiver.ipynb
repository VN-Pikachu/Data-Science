```python
import numpy as np
import matplotlib.pyplot as plt
```

# quiver

```python
x = np.linspace(-1,1,100)
y = np.linspace(-1,1,100)
xv, yv = np.meshgrid(x,y)
zv = xv * xv + yv * yv
```

```python
fig = plt.figure(figsize = (4,4))
ax = fig.add_axes([0,0,1,1])
ax.quiver(xv, yv, 2 * xv, 2 * yv)
```
