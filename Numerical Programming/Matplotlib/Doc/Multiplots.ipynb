````python
import numpy as np 
import matplotlib.pyplot as plt
x = np.linspace(-2 * np.pi, 2 * np.pi, 50)
````

# fig.add_subplot

````python
fig = plt.figure(figsize = (6,4))

ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

ax1.plot(x, np.sin(x), 'r')
ax2.plot(x, np.cos(x), ':')
ax3.plot(x, np.tanh(x), '^')
ax4.plot(x, np.exp(x), '--Hr')

plt.show()
````

# plt.subplot

# plt.GridSpec 

```python
plt.GridSpec(
    nrows,
    ncols,
    figure=None,
    left=None,
    bottom=None,
    right=None,
    top=None,
    wspace=None,
    hspace=None,
    width_ratios=None,
    height_ratios=None,
)```

**`plt.GridSpec()`** is equivalent to **`Figure.add_gridpec`**

````python
grid = plt.GridSpec(2,2, wspace = 0, hspace = 0) #equivalent: Figure.add_gridspec(2,2,hspace=0,wspace=0)
````

````python
fig = plt.figure()
ax1 = fig.add_subplot(grid[0,:], facecolor = 'r')
ax2 = fig.add_subplot(grid[1, 0], facecolor = 'k')
ax3 = fig.add_subplot(grid[1, 1], facecolor = 'b')

ax1.plot(x, np.sin(x))
ax2.plot(x, np.cos(x))
ax3.plot(x, np.tanh(x))
plt.show()
````

**`height_ratios`**, **`width_ratios`**

````python
#subplots of different sizes (using height_ratios and width_ratios)
fig = plt.figure()
grid = fig.add_gridspec(3,3,height_ratios = (1,2,3), width_ratios = (1,2,3))

for i in range(3):
    for j in range(3):
        ax = fig.add_subplot(grid[i,j])
        ax.set(xticks=[],yticks=[])
        if j == 0:
            ax.set_ylabel(f'height = {i + 1}')
        if i == 2:
            ax.set_xlabel(f'width = {j + 1}')
````

# plt.subplot2grid

return an axes class

```python
plt.subplot2grid(shape, loc, rowspan=1, colspan=1, fig=None, **kwargs)
```

````python
fig = plt.figure(linewidth = 2, edgecolor = 'teal', figsize = (10, 5))
ax1 = plt.subplot2grid((3,3), (0,0), rowspan = 2, colspan = 2, fig = fig)
ax2 = plt.subplot2grid((3,3), (0,2), rowspan = 3, fig = fig)
ax3 = plt.subplot2grid((3,3), (2,0), colspan = 2, fig = fig)

ax1.plot(x, np.sin(x))
ax2.plot(x, np.cos(x))
ax3.plot(x, np.tanh(x))
plt.show()
````

# plt.subplots_adjust

```python
plt.subplots_adjust(
    left=None,
    bottom=None,
    right=None,
    top=None,
    wspace=None,
    hspace=None,
)```

* **`wspace`**: the distance between Subplots (horizontally), relative to the size of the Subplot  
* **`hspace`**: the distance between Subplots (vertically), relative to the size of the Subplot 

````python
fig, ax = plt.subplots(2,2, sharex = True, sharey = True, figsize = (10,10), facecolor = 'grey', subplot_kw=dict(xticks=[],yticks=[]))

ax[0,0].plot(x, np.sin(x), 'r')
ax[0,1].plot(x, np.cos(x), ':')
ax[1,0].plot(x, np.tanh(x), '^')
ax[1,1].plot(x, np.exp(x), '--Hr')

ax[0,0].set_ylim(-1.1, 1.1)
plt.subplots_adjust(hspace = 0, wspace = 0)
plt.show()
````

**`top`**, **`bottom`**,**`left`**,**`right`**: The position of subplots, relative to the size of Figure

````python
fig, axes = plt.subplots(2,2, facecolor = 'gold', subplot_kw = dict(xticks=[], yticks=[]), figsize = (5,5))
fig.subplots_adjust(left = 0.25, right = 0.75, bottom = 0.25, top = 0.75)
````

# Layout Control

Reduce overlap between Axes: **`Figure.tight_layout()`**

````python
#without Figure.tight_layout()
fig = plt.figure()
grid = plt.GridSpec(2,2)

ax1 = fig.add_subplot(grid[0,0])
ax2 = fig.add_subplot(grid[0,1])
ax3 = fig.add_subplot(grid[1,0])
ax4 = fig.add_subplot(grid[1,1])

ax1.plot(x, np.sin(x))
ax2.plot(x, np.cos(x))
ax3.plot(x, np.tanh(x))
ax4.plot(x, 1 / np.tanh(x))
````

````python
#with Figure.tight_layout()
fig = plt.figure()
grid = plt.GridSpec(2,2)

ax1 = fig.add_subplot(grid[0,0])
ax2 = fig.add_subplot(grid[0,1])
ax3 = fig.add_subplot(grid[1,0])
ax4 = fig.add_subplot(grid[1,1])

ax1.plot(x, np.sin(x))
ax2.plot(x, np.cos(x))
ax3.plot(x, np.tanh(x))
ax4.plot(x, 1 / np.tanh(x))
fig.tight_layout()
````

Note that **`matplotlib.pyplot.tight_layout()`** will only adjust the subplot params when it is called.  
In order to perform this adjustment each time the figure is redrawn, you can call **`fig.set_tight_layout(True)`** or equivalently, set **`rcParams["figure.autolayout"]`** (default: False) to True.

````python
plt.rcParams['figure.autolayout'] = True
fig = plt.figure()
grid = fig.add_gridspec(2,2)

ax1 = fig.add_subplot(grid[0,0])
ax2 = fig.add_subplot(grid[0,1])
ax3 = fig.add_subplot(grid[1,0])
ax4 = fig.add_subplot(grid[1,1])

ax1.plot(x, np.sin(x))
ax2.plot(x, np.cos(x))
ax3.plot(x, np.tanh(x))
ax4.plot(x, 1 / np.tanh(x))
````
