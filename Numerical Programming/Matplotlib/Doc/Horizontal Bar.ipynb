```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(6)

scores = np.random.randint(30, 80, 5)
names = ['Tommy', 'John', 'Alan', 'Sophia', 'Peter']
vary = [3, 5, 2, 8, 6]
plt.barh(names, scores, xerr = vary, zorder = 3)
plt.title('How good do you score this time?')
plt.grid(True, linestyle = '--', color = 'gray', alpha = .5, linewidth = 1, zorder = -1)
plt.xlim(0, 100)
plt.axvline(x = 50)
plt.xlabel('Scores')
plt.show()
```
