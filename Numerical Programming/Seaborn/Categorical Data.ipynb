# Three different families of categorical plot

## 1. Categorical scatter plot

<b style='color:red'><u>Best Practice</u>:Use these plots when the size of the distribution is small</b>

### sns.stripplot (Axes-level function)

```python
sns.stripplot(
    x=None,
    y=None,
    hue=None,
    data=None,
    order=None,
    hue_order=None,
    jitter=True,
    dodge=False,
    orient=None,
    color=None,
    palette=None,
    size=5,
    edgecolor='gray',
    linewidth=0,
    ax=None,
    **kwargs,
)
```

````python
# help(sns.stripplot)
````

````python
#2 ways to plot data(like matplotlib)
sns.stripplot(x = 'day', y = 'total_bill', data = tips, hue = 'time')
#equivalent: sns.stripplot(x = tips['day'], y = tips['total_bill'], hue = tips['time'])
````

````python
#Jitter adds some random noise to the data. This parameter will adjust the positions along the categorical axis.
#jiter = False
sns.stripplot(x = tips['day'], y = tips['total_bill'], hue = tips['time'], jitter = False)
````

````python
#dodge: seperate different hue
fig, axes = plt.subplots(1,2)
sns.stripplot(x = 'sex', y = 'tip', data = tips, hue = 'smoker', dodge = False, ax = axes[0])
sns.stripplot(x = 'sex', y = 'tip', data = tips, hue = 'smoker', dodge = True, ax = axes[1])

axes[0].set_title('dodge = False')
axes[1].set_title('dodge = True')
````

````python
#customize
#set maker size to 4 (size = 4)
#set marker edgecolor to black (edgecolor = 'k')
#set marker edgewidth to 1(linewidth = 1)
#color palette to 'husl' (palette = 'husl')
sns.stripplot(x = 'sex', y = 'tip', data = tips, hue = 'smoker', size = 4, palette = 'husl', edgecolor = 'k', linewidth = 1)
````

````python
#customize
#set the width of line around the boxes to 5 (linewidth = 5)
#set the width of the box to 0.5 (width = 0.5)
#set the marker size of outliers to 20 (fliersize = 20)
sns.boxplot(x = 'total_bill', y = 'sex', data = tips, hue = 'smoker', linewidth = 5, width = 0.5, fliersize = 20)
````

````python
# help(sns.catplot)
````

````python
#let's try to boxplot (categorical distribution plot) using catplot
sns.catplot(x = 'sex', y = 'total_bill', hue = 'smoker', data = tips, kind = 'box')
````

````python
#barplot (categorical estimate plot) using catplot, using median as a measurement of central tendency
#order days by Thur, Fri, Sat, Sun
sns.catplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips, kind = 'bar', estimator = np.median, order = ('Thur', 'Fri', 'Sat', 'Sun'))
````

In factorplot, the data is plotted on a **facet grid**.  
**Facet grid** forms a matrix of panels defined by row and column by dividing the variables.  
Due of panels, a single plot looks like multiple plots. It is very helpful to analyze all combinations in two discrete variables.

````python
#plot on Facet
sns.catplot(x = 'day', y = 'total_bill', hue = 'time', col = 'sex', row = 'smoker', data = tips, kind = 'violin')
````

````python
titanic = pd.read_csv('./titanic.csv')
titanic.head()
````

````python
sns.catplot(x = 'Sex', y = 'Fare', col = 'Pclass', data = titanic, kind = 'box')
````

````python
#CUSTOMIZE
#col_order: p_class = 3, p_class = 2, p_class = 1 (col_order = (3,2,1))
#each facet of size 3 x 3 (height = 3)
#set aspect ratio of each facet (width / height ratio) to 3 : 2 (aspect = 1.5)
#using palette = 'husl'
sns.catplot(x = 'Survived', hue = 'Sex', col = 'Pclass', data = titanic, kind = 'count', 
            col_order = (3,2,1), height = 3, aspect = 1.5, palette = 'husl')
````

````python
tips = sns.load_dataset('tips')
sns.catplot(data = tips, x = 'day', y = 'total_bill', col = 'time', hue = 'smoker', kind = 'swarm')
````

For further customization of the plot, you can use the methods on the FacetGrid object that it returns:

````python
titanic = sns.load_dataset('titanic')
g = sns.catplot(x="fare", y="survived", row="class",
                kind="box", orient="h", height=1.5, aspect=4,
                data=titanic.query("fare > 0"))

g.set(xscale = 'log')
````
