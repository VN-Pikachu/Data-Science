````python
pokemon = pd.read_csv('./pokemon.csv', index_col = 0)
pokemon['Legendary'] = pokemon['Legendary'].apply(lambda v: 'Legendary' if v else 'Normal')
pokemon.head(5)
````

````python
stats = pokemon.loc[:, ['HP', 'Attack','Defense', 'Legendary']]
stats.head()
````

Datasets under real-time study contain many variables.  
In such cases, the relation between each and every variable should be analyzed.
To plot multiple pairwise bivariate distributions in a dataset, you can use the **`pairplot()`** function. 

# sns.pairplot

```python
sns.pairplot(
    data,
    hue=None,
    hue_order=None,
    palette=None,
    vars=None,
    x_vars=None,
    y_vars=None,
    kind='scatter',
    diag_kind='auto',
    markers=None,
    height=2.5,
    aspect=1,
    dropna=True,
    plot_kws=None,
    diag_kws=None,
    grid_kws=None,
    size=None,
)
```

<table>
    <tr><td><b>hue</b></td><td> Variable in data to map plot aspects to different colors.</td></tr>
    <tr><td><b>palette</b></td><td> Set of colors for mapping the hue variable</td></tr>
    <tr><td><b>kind</b></td><td> Kind of plot for the non-identity relationships. {‘scatter’, ‘reg’}</td></tr>
    <tr><td><b>diag_kind</b></td><td> Kind of plot for the diagonal subplots. {‘hist’, ‘kde’}</td></tr>
</table>

````python
help(sns.pairplot)
````

````python

#hue = 'Legendary' : set different for Legendary and not legendary pokemon
sns.pairplot(stats, hue = 'Legendary', kind = 'reg', diag_kind = 'kde', palette = 'rocket_r', markers = ['s', 'o'])
````

````python
#select a subset column('HP', 'Defense', 'Sp. Def') from data to plot:
#data = pokemon, vars = ['HP', 'Defense', 'Sp. Def']

#set the size of each facet to 1 x 1 (height = 1)

sns.pairplot(pokemon, vars = ['HP', 'Defense', 'Sp. Def'], diag_kind = 'kde', kind = 'scatter', hue = 'Legendary', height = 1)
````

````python
#select x axis to be HP, Defense
#select y axis to by Attack, Speed

sns.pairplot(pokemon, x_vars = ['HP', 'Defense'], y_vars = ['Attack', 'Speed'], hue = 'Generation', diag_kind = 'kde', kind = 'reg', height = 4)
````
