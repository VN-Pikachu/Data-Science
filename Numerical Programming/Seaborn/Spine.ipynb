````python
import seaborn as sns;sns.set()
import numpy as np
import matplotlib.pyplot as plt
````

````python
x = np.linspace(-10,10, 100)
````

# Despine

**`sns.despine()`**: removing axes spines, default remove top and right spines

```python
sns.despine(
    fig=None,
    ax=None,
    top=True,
    right=True,
    left=False,
    bottom=False,
    offset=None,
    trim=False,
)```

````python
sns.set_style('white')
fig, ax = plt.subplots()
ax.plot(x, np.sinc(x))

#let's remove top and bottom spine
sns.despine(fig = fig, ax = ax, top = False, bottom = False, left = True, right = True)
````

---
For a figure-level function, use **`FacetGrid.depsine`**

````python
tips = sns.load_dataset('tips')
with sns.axes_style('whitegrid'):
    g = sns.relplot(data = tips, x = 'total_bill', y = 'tip', col = 'day', col_wrap = 2)
    g.despine(left=True, bottom=True)

````

## Offset and trim

````python
# read document for its usage
sns.boxplot(data = tips, x = 'day', y = 'total_bill', hue = 'smoker', palette=['m', 'g'])
sns.despine(trim = True, offset = 10)
````
