```python
sns.set_style('whitegrid', rc = {'xtick.color' : 'gold', 'ytick.color': 'teal', 'grid.linestyle': '--'})

x = np.linspace(-10, 10, 100)
y = np.sinc(x)
fig, axes = plt.subplots(1,2, figsize = (8,4))
axes[0].plot(x, y, 'r')
axes[1].plot(x, np.sin(x), 'k')
sns.despine(ax = axes[0])
sns.despine(ax = axes[1], top = True, left = True, bottom = False, right = False)

axes[0].set_title('Sinc')
axes[1].set_title('Sine')
fig.tight_layout()
```
