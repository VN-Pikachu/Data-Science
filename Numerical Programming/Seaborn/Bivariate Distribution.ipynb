Bivariate Distribution is used to determine the relation between two variables. This mainly deals with relationship between two variables and how one variable is behaving with respect to the other.

The best way to analyze Bivariate Distribution in seaborn is by using the **`jointplot()`** function.

Jointplot creates a multi-panel figure that projects the bivariate relationship between two variables and also the univariate distribution of each variable on separate axes.

# sns.jointplot

```python
sns.jointplot(
    x,
    y,
    data=None,
    kind='scatter',
    stat_func=None,
    color=None,
    height=6,
    ratio=5,
    space=0.2,
    dropna=True,
    xlim=None,
    ylim=None,
    joint_kws=None,
    marginal_kws=None,
    annot_kws=None,
    **kwargs,
)
```

````python
# help(sns.jointplot)
````

`jointplot()` is a convenient interface to the `JointGrid` class, which offeres more flexibility when used directly:



````python
g = sns.JointGrid(data = iris, x = 'sepal_length', y = 'sepal_width')
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
````

2 ways to visualize: (like **`matplotlib`**):  
**`sns.jointplot(x = df['Attack'], y = df['Defense'])`**  
or equivalent  
**`sns.jointplot(x = 'Atack', y = 'Defense', data = df)`**

````python
#scatter plot
sns.jointplot(x = pokemon['Attack'], y = pokemon['Defense'], kind = 'scatter')
````

````python
#equivalent 
sns.jointplot(x = 'Attack', y = 'Defense', data = pokemon, kind = 'scatter')
````

````python
#hex plot (kind = 'hex')
sns.jointplot(x = 'HP', y = 'Defense', data = pokemon, kind = 'hex')
````

````python
#kde plot (kind = 'kde')
sns.jointplot(x = pokemon['Attack'], y = pokemon['Speed'], kind = 'kde')
````

````python
#linear regression fit (kind = 'reg')
sns.jointplot(x = 'Attack', y = 'Defense', data = pokemon, kind = 'reg')
````

````python
#residual point: y_target and y_predict
y_target = np.random.randint(1, 10, 30)
y_predict = np.random.randint(1, 10, 30)
print(y_target)
print(y_predict)
sns.jointplot(x = y_target, y = y_predict, data = pokemon, kind = 'resid')
````

````python
#customize:
#set marker color to teal(color = 'teal')
#figure size to 8x8(height = 8)
#ratio of joint axes to marginal axes: 1 (ratio = 1)
#space bewteen join axes and marginal axes: 1 (space = 1)
sns.jointplot(x = 'Attack', y = 'Speed', data = pokemon, color = 'teal', height = 8, ratio = 1, space = 1)
````

# `sns.displot` (figure-level function)

````python
diamonds = sns.load_dataset('diamonds')
diamonds.head()
````

````python
penguins = sns.load_dataset('penguins')
penguins.head()
````

`geom_bin2d`

````python
# default kind = 'hist'
sns.displot(data = penguins, x = 'bill_length_mm', y = 'bill_depth_mm', hue = 'species')
````

`geom_density2d`

````python
sns.displot(data = penguins, x = 'bill_length_mm', y = 'bill_depth_mm', kind = 'kde', hue = 'species')
````

Just as with univariate plots, the choice of bin size or smoothing bandwidth will determine how well the plot represents the underlying bivariate distribution. The same parameters apply, but they can be tuned for each variable by passing a pair of values:

````python
sns.displot(data = penguins, x = 'bill_length_mm', y = 'bill_depth_mm', hue = 'species', binwidth = (5, 2))
````

To aid interpretation of the heatmap, add a colorbar to show the mapping between counts and color intensity:

````python
sns.displot(data = penguins, x = 'bill_length_mm', y = 'bill_depth_mm', cbar = True)
````

control contour levels

````python
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", kind="kde", levels=[.01, .05, .1, .8])
````

Categorical vs Continuous

````python
sns.displot(data = diamonds, x = 'clarity', y = 'price', cbar = True)
````

Categorical vs Categorical

````python
sns.displot(data = diamonds, x = 'clarity', y = 'color', cbar = True)
````
