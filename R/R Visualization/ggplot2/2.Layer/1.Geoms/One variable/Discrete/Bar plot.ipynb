# Bar charts

There are two types of bar charts: **`geom_bar()`** and **`geom_col()`**. **`geom_bar()`** makes the height of the bar proportional to the number of cases in each group (or if the weight aesthetic is supplied, the sum of the weights). If you want the heights of the bars to represent values in the data, use `geom_col()` instead. `geom_bar()` uses `stat_count()` by default: it counts the number of cases at each x position. `geom_col()` uses `stat_identity()`: it leaves the data as is.

```R
geom_bar(
  mapping = NULL,
  data = NULL,
  stat = "count",
  position = "stack",
  ...,
  width = NULL,
  na.rm = FALSE,
  orientation = NA,
  show.legend = NA,
  inherit.aes = TRUE
)

geom_col(
  mapping = NULL,
  data = NULL,
  position = "stack",
  ...,
  width = NULL,
  na.rm = FALSE,
  show.legend = NA,
  inherit.aes = TRUE
)

stat_count(
  mapping = NULL,
  data = NULL,
  geom = "bar",
  position = "stack",
  ...,
  width = NULL,
  na.rm = FALSE,
  orientation = NA,
  show.legend = NA,
  inherit.aes = TRUE
)
```

<b style = 'color:red'>NOTE:<br>
By default, multiple bars occupying the same x position will be stacked atop one another by `position_stack()`. If you want them to be dodged `side-to-side`, use `position_dodge()` or `position_dodge2()`. Finally, `position_fill()` shows relative proportions at each x by stacking the bars and then standardising each bar to have the same height.
</b>

# Aesthetic

**`geom_bar()`** and **`geom_col`** understands the following aesthetics (required aesthetics are in bold):

**x**

**y**

alpha

colour

fill

group

linetype

size

**`stat_count()`** understands the following aesthetics (required aesthetics are in bold):

**x** or **y**

group

weight

# Example

````r
#The number of people in each pclass
ggplot(titanic, aes(Pclass)) +
geom_bar()
````

````r
#flipping orientation
ggplot(titanic, aes(y = Pclass)) +
geom_bar()
````

````r
# Bar charts are automatically stacked when multiple bars are placed
# at the same location. The order of the fill is designed to match
# the legend

#number of people in each Pclass, hue = 'Sex'
ggplot(titanic, aes(x = Pclass, fill = Sex)) +
geom_bar()
````

````r
# If you need to flip the order (because you've flipped the orientation)
# call position_stack() explicitly:
ggplot(titanic, aes(y = Pclass, fill = Sex)) +
geom_bar(position = position_stack(reverse = TRUE))
````

````r
#side by side bar: position = 'dodge'
ggplot(titanic, aes(Pclass, fill = Sex)) +
geom_bar(position = 'dodge')
````

````r
#side by side bar with space between bars in the same group
#position = 'dodge2'
ggplot(titanic, aes(Pclass, fill = Sex)) +
geom_bar(position = 'dodge2')
````

````r
#Customization
ggplot(titanic, aes(Pclass)) +
geom_bar(fill = 'blue',       #facecolor = 'blue',
         color = 'gold',      #edgecolor = 'gold',
         linetype = 'dashed', #linestyle = '--'
         size = 3             #linewidth = 3
        )
````

````r
tanks <- data.frame(
    Name = c('ZBL09', 'Stingray', 'Leclerc'),
    Price = c(64000L, 22000L, 48000L)
)
tanks
````

````r
#display the data: stat = 'identity'
ggplot(tanks, aes(Name, Price)) +
geom_bar(stat = 'identity')
````

````r
#Equivalently, using geom_col
ggplot(tanks, aes(Name, Price)) +
geom_col()
````
