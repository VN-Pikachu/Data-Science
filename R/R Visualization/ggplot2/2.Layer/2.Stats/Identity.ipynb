# Leave data as is

```R
stat_identity(
  mapping = NULL,
  data = NULL,
  geom = "point",
  position = "identity",
  ...,
  show.legend = NA,
  inherit.aes = TRUE
)
```

# Examples

````r
p <- ggplot(mtcars, aes(wt, mpg))
p + stat_identity() #equivalent: geom_point(stat = 'identity')
````

````r
clan <- data.frame(names <- LETTERS[1:3], values <- c(5, 2, 3))

ggplot(clan, aes(names, values)) +
stat_identity(geom = 'bar')   #equivalent: geom_bar(stat = 'identity'), geom_col()
````
