# Examples

Values calculated by the stat is available with the `after_stat()` function 
inside `aes()`. You can do all sorts of computations inside that.

````r
library(ggplot2)
````

**`Geom_histogram`** return `count`

````r
#normalized histogram:

ggplot(mpg, aes(displ)) +
geom_histogram(aes(y = after_stat(count / sum(count))))
````

````r
# Use a transparent version of colour for fill
ggplot(mpg, aes(class, hwy)) +
  geom_boxplot(aes(colour = class, fill = after_scale(alpha(colour, 0.4)))) #se ?alpha for more information
                                                                            
````

````r
# Use stage to modify the scaled fill
ggplot(mpg, aes(class, hwy)) +
  geom_boxplot(aes(fill = stage(class, after_scale = alpha(fill, 0.4))))
````
