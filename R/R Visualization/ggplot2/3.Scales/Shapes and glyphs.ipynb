# Scales for shapes, aka glyphs

**`scale_shape()`** maps discrete variables to six easily discernible shapes. If you have more than six levels, you will get a warning message, and the seventh and subsequence levels will not appear on the plot. Use `scale_shape_manual()` to supply your own values. You can not map a continuous variable to shape unless `scale_shape_binned()` is used. Still, as shape has no inherent order, this use is not advised.

```R
scale_shape(..., solid = TRUE)

scale_shape_binned(..., solid = TRUE)
```

# Examples

````r
library(ggplot2)
````

````r
# Show a list of available shapes
df_shapes <- data.frame(shape = 0:24)
ggplot(df_shapes, aes(0, 0, shape = shape)) +
  geom_point(aes(shape = shape), size = 5, fill = 'red') +
  scale_shape_identity() +
  facet_wrap(~shape) +
  theme_void()
````

````r
head(mtcars)
````

````r
plot <- ggplot(mtcars, aes(disp, mpg, shape = factor(carb))) + geom_point()
plot
````

````r
#show solid shape (solid shape are shapes that can be filled by color)
plot + scale_shape(solid = TRUE) #default
````

````r
#do not show solid shape
plot + scale_shape(solid = FALSE)
````

````r
#set label for scale (in this case, the title of legend)

plot + scale_shape('Carbin')
````
