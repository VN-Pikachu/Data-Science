# Create your own discrete scale

These functions allow you to specify your own set of mappings from levels in the data to aesthetic values.

```R
scale_colour_manual(..., values, aesthetics = "colour", breaks = waiver())

scale_fill_manual(..., values, aesthetics = "fill", breaks = waiver())

scale_size_manual(..., values, breaks = waiver())

scale_shape_manual(..., values, breaks = waiver())

scale_linetype_manual(..., values, breaks = waiver())

scale_alpha_manual(..., values, breaks = waiver())

scale_discrete_manual(aesthetics, ..., values, breaks = waiver())
```

# Examples

````r
setwd('C:/Users/dell/PycharmProjects/MachineLearning/Pandas/datasets')
tips <- read.csv('./tips.csv')
````

````r
p <- ggplot(mtcars, aes(mpg, wt)) +
  geom_point(aes(colour = factor(cyl)))

p + scale_color_manual(values = c('red', 'green', 'blue'))
````

````r
# It's recommended to use a named vector
p + scale_color_manual(values =  c("8" = "red", "4" = "blue", "6" = "darkgreen", "10" = "orange"))
````

````r
# You can set color and fill aesthetics at the same time
ggplot(
  mtcars,
  aes(mpg, wt, colour = factor(cyl), fill = factor(cyl))
) +
  geom_point(shape = 21, alpha = 0.5, size = 2) + 
scale_color_manual(values = c('red', 'green', 'blue'), aesthetics = c('fill', 'color'))
````

````r
# As with other scales you can use breaks to control the appearance
# of the legend.

ggplot(tips, aes(total_bill, tip, color = sex)) +
geom_point() +
scale_color_manual(values = c('green', 'yellow'),
                   #Mapping legend text: Male --> Boy, Female --> Girl
                   breaks = c('Male', 'Female'),
                   labels = c('Boy', 'Girl'))
````

````r
# And limits to control the possible values of the scale
ggplot(tips, aes(total_bill, tip, color = day)) +
geom_point() + 
#only show values on Sunday and Friday
scale_color_manual(limits = c('Fri', 'Sun'), values = c('gold', 'blue'))
````
