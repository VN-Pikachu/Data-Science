# Binned gradient colour scales

**`scale_*_steps`** creates a two colour binned gradient (low-high)  
**`scale_*_steps2`** creates a diverging binned colour gradient (low-mid-high)  
**`and scale_*_stepsn`** creates a n-colour binned gradient.  
These scales are binned variants of the gradient scale family and works in the same way.

```R
scale_colour_steps(
  ...,
  low = "#132B43",
  high = "#56B1F7",
  space = "Lab",
  na.value = "grey50",
  guide = "coloursteps",
  aesthetics = "colour"
)

scale_colour_steps2(
  ...,
  low = muted("red"),
  mid = "white",
  high = muted("blue"),
  midpoint = 0,
  space = "Lab",
  na.value = "grey50",
  guide = "coloursteps",
  aesthetics = "colour"
)

scale_colour_stepsn(
  ...,
  colours,
  values = NULL,
  space = "Lab",
  na.value = "grey50",
  guide = "coloursteps",
  aesthetics = "colour",
  colors
)

scale_fill_steps(
  ...,
  low = "#132B43",
  high = "#56B1F7",
  space = "Lab",
  na.value = "grey50",
  guide = "coloursteps",
  aesthetics = "fill"
)

scale_fill_steps2(
  ...,
  low = muted("red"),
  mid = "white",
  high = muted("blue"),
  midpoint = 0,
  space = "Lab",
  na.value = "grey50",
  guide = "coloursteps",
  aesthetics = "fill"
)

scale_fill_stepsn(
  ...,
  colours,
  values = NULL,
  space = "Lab",
  na.value = "grey50",
  guide = "coloursteps",
  aesthetics = "fill",
  colors
)
```

````r
plot <- ggplot(faithfuld, aes(waiting, eruptions, fill = density)) + geom_raster()
````

````r
#gradient
plot + scale_fill_gradient() + labs(title = 'gradient')
#binned gradient: convert continuous values to discrete values by using binning (like histogram)
plot + scale_fill_steps() + labs(title = 'binned gradient')
````

````r
#diverging binned gradient
ggplot(faithfuld, aes(waiting, eruptions, fill = density - 0.02)) + 
geom_raster() + 
scale_fill_steps2()
````

````r
# Define your own colour ramp to extract binned colours from

plot + scale_fill_stepsn(colours = terrain.colors(10))
````
