# Continuous and binned colour scales

```R
scale_colour_continuous(
  ...,
  type = getOption("ggplot2.continuous.colour", default = "gradient")
)

scale_fill_continuous(
  ...,
  type = getOption("ggplot2.continuous.fill", default = "gradient")
)
```
**Arguments**   
`...`	
Additional parameters passed on to the scale type

`type`	
One of the following:

* "gradient" (the default)

* "viridis"

* A function that returns a continuous colour scale.

# Example

````r
library(ggplot2)
````

````r
head(faithfuld)
````

````r
plot <- ggplot(faithfuld, aes(waiting, eruptions, fill = density)) + geom_raster()
plot
````

````r
plot + scale_fill_continuous(type = 'gradient') 
#equivalent
plot + scale_fill_gradient()
````

````r
plot + scale_fill_continuous(type = 'viridis')
#equivalent
plot + scale_fill_viridis_c()
````
