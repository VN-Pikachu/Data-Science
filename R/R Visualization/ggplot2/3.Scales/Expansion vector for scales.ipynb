The expansion vectors are used to add some space between the data and the axes.

```R
expansion(mult = 0, add = 0)

expand_scale(mult = 0, add = 0)
```

# Example

````r
bars <- data.frame(x = 1:3, y = 1)
bars
````

````r

plot <- ggplot(bars, aes(x, y)) + geom_col()
plot
````

````r
# No space below the bars but 50% above them
plot + scale_y_continuous(expand = expansion(mul = c(0, 0.5)))
````

````r
# Add 3 units of space on the left and right of the data
plot + 
scale_x_continuous(expand = expansion(add = 3))
````

````r
# Reproduce the default range expansion used
# when the 'expand' argument is not specified
plot + 
scale_y_continuous(expand = expansion(mul = .05)) 
````
