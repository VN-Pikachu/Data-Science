# Position scales for discrete data

**`scale_x_discrete()`** and **`scale_y_discrete()`** are used to set the values for discrete x and y scale aesthetics.  
For simple manipulation of scale labels and limits, you may wish to use `labs()` and `lims()` instead.

```R
scale_x_discrete(..., expand = waiver(), guide = waiver(), position = "bottom")

scale_y_discrete(..., expand = waiver(), guide = waiver(), position = "left")
```

````r
tips <- read.csv('./tips.csv')
head(tips)
````

````r
# The discrete position scale is added automatically whenever you
# have a discrete position.
plot <- ggplot(tips, aes(day, total_bill, fill = sex)) + geom_boxplot()
plot
````

````r
#set label for the scale (in this case, it is x-axis label)
plot + scale_x_discrete('Day of the week')
````

````r

# Use limits to adjust the which levels (and in what order)
# are displayed
plot + scale_x_discrete(limits = c('Fri', 'Sun')) #Display only the result of 'Fri', 'Sun'
````

````r
#Mapping discrete xticklabels by using a mapping object
plot + scale_x_discrete(labels = c(Fri = 'Friday', Sat = 'Saturday', Sun = 'Sunday', Thur = 'Thursday'))
````

````r
#You can also use a function to map
decorate <- function(labels) paste('xXx-', labels, '-xXx')

plot + scale_x_discrete(labels = decorate)
````

````r
#turn off x labels
plot + 
scale_x_discrete(labels = NULL)
````

````r
#turn off xticks, xticklabels
plot + 
scale_x_discrete(breaks = NULL)
````
