```r
library(ggplot2)
```

**`hjust`**: `c('left', 'center', 'right')` or numeric vector
**`vjust`**: `c('bottom', 'center', 'left')` or numeric vector

```r
point <- data.frame(x = 0, y = 0)

plot <- ggplot(point, aes(x, y)) + geom_point(color = 'red')
plot
```

`just = 1` means `horizontal alignment = right`. The rightmost edge of the text will be place at the `position` of the text (the red point)

```r
plot + 
geom_text(label = 'Word', hjust = 1) 
```

Left alignment

```r
plot +
geom_text(label = 'World',, hjust = 0)
```

horizontal center

```r
plot +
geom_text(label = 'World', hjust = 0.5) #default
```

vertical alignment: Top  
The top edge of the text is place at the position of the text (the red dot)

```r
plot + 
geom_text(label = 'World', vjust = 1)
```

vetical alignment: bottom

```r
plot + 
geom_text(label = 'World', vjust = 0)
```
