```r
library(ggplot2)
data(iris) #loading iris dataset
```

Aesthetic mappings describe how variables in the data are mapped to visual properties (aesthetics) of geoms. Aesthetic mappings can be set in **`ggplot()**` and in individual layers.

Set aesthetic mapping in `ggplot`

```r
#The data is the `Iris` dataset
#The aesthetic mapping:
#    + The x axis is the data of column `Sepal.Length`
#    + The y axis is the data of column `Sepal.Width`


ggplot(iris, aes(x = Sepal.Length, y = Sepal.Width)) + 
geom_point() #using point plot
```

We can aslo set aesthetic mapping in indivisual layer

```r
ggplot(iris) +
geom_point(aes(Sepal.Length, Sepal.Width))
```

aesthemic mapping can also be used to customize:

```r
fig <- ggplot(iris, aes(x = Sepal.Length, 
                        y = Sepal.Width, 
                        col = Species, #specify color of points match the value of 'Species',
                        shape = Species #specify marker for each `Species`
                       )) +
geom_point() +
theme_classic() + #chosing a theme: classic
labs(x = 'Sepal length', y = 'Sepal width', title = 'Iris dataset') #labels
fig
```

Use aesthetic mapping for each layer

```r
ggplot(iris, aes(Sepal.Length, Sepal.Width)) + 
geom_point(aes(color = Species)) #use aesthetic mapping within a layer `geom_point`
```
