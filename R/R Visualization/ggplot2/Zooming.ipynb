```r
library(ggplot2)
```

<img src = './ggplot2_zooming.png'/>

```r
r <- 5
n <- 9
theta <- seq(0, 2 * pi, length.out = n)

x <- r * cos(theta)
y <- r * sin(theta)

v <- data.frame(x, y)

plot <- ggplot(v, aes(x, y)) + geom_polygon(fill = 'white', size = 3, color = 'black')
plot
```

# Zooming without clipping

```r
plot + coord_cartesian(xlim = c(0, 5))
```

# Zooming with clipping

```r
#before plotting, remove values that are outside of the limit
plot + xlim(0, 5)
```

```r
plot + scale_x_continuous(limits = c(0, 5))
```
