# Continuous colour bar guide

Colour bar guide shows continuous colour scales mapped onto values. Colour bar is available with `scale_fill` and `scale_colour`

```R
guide_colorbar(
  title = waiver(),
  title.position = NULL,
  title.theme = NULL,
  title.hjust = NULL,
  title.vjust = NULL,
  label = TRUE,
  label.position = NULL,
  label.theme = NULL,
  label.hjust = NULL,
  label.vjust = NULL,
  barwidth = NULL,
  barheight = NULL,
  nbin = 300,
  raster = TRUE,
  frame.colour = NULL,
  frame.linewidth = 0.5,
  frame.linetype = 1,
  ticks = TRUE,
  ticks.colour = "white",
  ticks.linewidth = 0.5,
  draw.ulim = TRUE,
  draw.llim = TRUE,
  direction = NULL,
  default.unit = "line",
  reverse = FALSE,
  order = 0,
  available_aes = c("colour", "color", "fill"),
  ...
)
```

````r
args(theme)
````

# Examples

````r
plot <- ggplot(faithfuld, aes(waiting, eruptions, fill = density)) + geom_raster()
plot
````

````r
#Basic form
plot +
scale_fill_continuous(guide = guide_colorbar()) #equivalent, scale_fill_continuous(guide = 'colorbar')
````

<hr>

## Title

````r
#Turn off the title
plot + 
scale_fill_continuous(guide = guide_colorbar(title = NULL))
````

````r
#set title name
plot + 
scale_fill_continuous(guide = guide_colorbar(title = 'DENSITY MASS'))
````

````r
#Set the position of the title
plot + 
scale_fill_continuous(guide = guide_colorbar(title.position = 'left')) #left, right, bottom, top
````

````r
#hjust, vjust
plot + 
scale_fill_continuous(guide = guide_colorbar(title.hjust = 0, label.position = 'left')) #left alignment
````

````r
#Customize the title (also via: theme(legend.title = element_text(...)))
plot + 
scale_fill_continuous(guide = guide_colorbar(title.theme = element_text(color = 'red', face = 'bold')))
````

## Bar

````r
#control style
#bar size
plot + scale_fill_continuous(guide = guide_colorbar(barheight = 2, barwidth = 4))
````

## Label

````r
#No label
plot +
scale_fill_continuous(guide = guide_colorbar(label = FALSE))
````

````r
#set the position of labels
plot +
scale_fill_continuous(guide = guide_colorbar(label.position = 'left'))
````

````r
#label theme
plot + 
scale_fill_continuous(guide = guide_colorbar(label.theme = element_text(face = 'bold', color = 'red')))
````

````r
#hjust, vjust for label
plot + 
scale_fill_continuous(guide = guide_colorbar(labels.hjust = 1)) #right alignment
````

## Ticks

````r
#No tick marks
plot + 
scale_fill_continuous(guide = guide_colorbar(ticks = FALSE))
````

````r
#customize ticks
plot + 
scale_fill_continuous(guide = guide_colorbar(ticks.colour = 'red', ticks.linewidth = 2))
````

# Bins

````r
#small the number of bins
plot + scale_fill_continuous(guide = guide_colorbar(nbin = 2))
#large number of bins
plot + scale_fill_continuous(guide = guide_colorbar(nbin = 100))
````

## Direction of the guide

````r
plot + 
scale_fill_continuous(guide = guide_colorbar(direction = 'horizontal'))
````

# Reverse

````r
plot + 
scale_fill_continuous(guide = guide_colorbar(reverse = TRUE))
````
