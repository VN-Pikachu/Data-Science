# Legend Guides


```R
guide_legend(
  title = waiver(),
  title.position = NULL,
  title.theme = NULL,
  title.hjust = NULL,
  title.vjust = NULL,
  label = TRUE,
  label.position = NULL,
  label.theme = NULL,
  label.hjust = NULL,
  label.vjust = NULL,
  keywidth = NULL,
  keyheight = NULL,
  direction = NULL,
  default.unit = "line",
  override.aes = list(),
  nrow = NULL,
  ncol = NULL,
  byrow = FALSE,
  reverse = FALSE,
  order = 0,
  ...
)
```

````r
plot <- ggplot(iris, aes(Sepal.Length, Sepal.Width, color = Species)) + geom_point()
plot
````

## Title, Labels, Reverse, Direction

# Key

````r
#width and height
plot + 
scale_color_discrete(guide = guide_legend(keywidth = 4, keyheight = 2))
````

# nrow, ncol

````r
plot + 
scale_color_discrete(guide = guide_legend(ncol = 3))
````

````r
#go from top to bottom
plot + 
scale_color_discrete(guide = guide_legend(byrow = FALSE, ncol = 2))
````

````r
#Go from left to right
plot + 
scale_color_discrete(guide = guide_legend(byrow = TRUE, ncol = 2))
````

# override.aes

````r
# override.aes overwrites the alpha
ggplot(iris, aes(Sepal.Length, Sepal.Width, color = Species)) + 
geom_point(alpha = .3) +
scale_color_discrete(guide = guide_legend(override.aes = list(alpha = 1)))
````
