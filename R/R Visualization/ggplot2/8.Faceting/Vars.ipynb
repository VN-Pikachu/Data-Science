# Quote faceting variables

Just like `aes()`, `vars()` is a **quoting function** that takes inputs to be evaluated in the context of a dataset. These inputs can be:

* variable names

* complex expressions

In both cases, the results (the vectors that the variable represents or the results of the expressions) are used to form faceting groups.

```R
vars(...)
```
**Arguments**
`...`	
Variables or expressions automatically quoted. These are evaluated in the context of the data to form faceting groups. Can be named (the names are passed to a labeller).

````r
tips <- read.csv('./tips.csv')
head(tips)
````

````r
ggplot(tips, aes(total_bill, tip, color = smoker)) +
geom_point() +
facet_wrap(vars(day))
````

Equivalently, you can also use **`facet_wrap(~day)`**, instead of using **`facet_wrap(vars(day))`**

````r
ggplot(tips, aes(total_bill, tip, color = smoker)) +
geom_point() +
facet_wrap(~day)
````

<hr>

````r
ggplot(tips, aes(total_bill, tip)) +
geom_point() +
facet_grid(vars(x = sex, y = smoker))
````

````r
ggplot(tips, aes(total_bill, tip)) +
geom_point() +
facet_grid(sex ~ smoker) #row are sex, column are smoker
````

<hr>

````r
# vars() makes it easy to pass variables from wrapper functions:
wrap_by <- function(...) {
    facet_wrap(vars(...))
}

plot <- ggplot(tips, aes(total_bill, tip)) + geom_point()

plot + wrap_by(day)

plot + wrap_by(day, time)
````
