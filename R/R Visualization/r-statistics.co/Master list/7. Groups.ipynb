# Hierarchical Dendrogram

```r
library(ggplot2)
library(ggdendro)
theme_set(theme_bw())

hc <- hclust(dist(USArrests), "ave")  # hierarchical clustering

# plot
ggdendrogram(hc, rotate = TRUE, size = 2)
```

# Cluster

It is possible to show the distinct clusters or groups using **`geom_encircle()`**

```r
library(ggalt)
```

```r
ggplot() + 
geom_point(data = iris, 
           mapping = aes(Sepal.Length, Sepal.Width, color = Species)) +
geom_encircle(data = iris %>% filter(Species == 'setosa'),
              mapping = aes(Sepal.Length, Sepal.Width), 
              color = 'tomato') +
geom_encircle(data = iris %>% filter(Species == 'versicolor'),
              mapping = aes(Sepal.Length, Sepal.Width), 
              color = 'green') +
geom_encircle(data = iris %>% filter(Species == 'virginica'),
              mapping = aes(Sepal.Length, Sepal.Width), 
              color = 'blue')
```
