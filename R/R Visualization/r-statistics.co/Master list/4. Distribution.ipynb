# Density Plot

````r
diamonds %>%
ggplot(aes(price, fill = cut)) + 
geom_density(alpha = .5) +
theme_classic()
````

# Box plot (Optional: dot plot)

````r
diamonds %>%
ggplot(aes(cut, price)) + 
geom_boxplot(fill = 'tomato') +
theme_classic()
````

# Tufte Plot

Tufte box plot, provided by ggthemes package is inspired by the works of Edward Tufte. Tuftes Box plot is just a box plot made minimal and visually appealing.

````r
diamonds %>%
ggplot(aes(cut, price)) +
geom_tufteboxplot(size = 1) +
theme_tufte()
````

# Violin Plot

````r
diamonds %>%
ggplot(aes(cut, price)) + 
geom_violin(fill = 'tomato') +
theme_classic()
````

'facet_share' uses [facet_wrap()] to build two panels with a shared axis.
```R
facet_share(
  facets,
  scales = "fixed",
  reverse_num = FALSE,     #Used when passing on flipped data (times -1) for the second (right/bottom) panel. 
                           #If 'TRUE', this will multiply the axis labels for that panel by -1.
  shrink = TRUE,
  labeller = "label_value",

```

Population pyramids offer a unique way of visualizing how much population or what percentage of population fall under a certain category. 

````r
df <- data.frame(
  Population = c(5, 8.7, 16.7, 24.8, 38, -4.6, -6.4, -16.1, -39.6, -55.3),
  Gender = c("Male", "Male", "Male", "Male", "Male", "Female", "Female", "Female", "Female", "Female"),
  AgeBand = c("65-69", "70-74", "75-79", "80-84", "85+", "65-69", "70-74", "75-79", "80-84", "85+")
)
df
````

````r

df %>%
ggplot(aes(x = AgeBand, y = Population, fill = Gender)) +
geom_col() +
facet_share(~Gender, scales = 'free', reverse_num = T) + 
coord_flip() +
theme(panel.background = element_rect(fill = 'white'),
      panel.grid = element_line(color = 'grey80'),
      strip.background = element_blank(),
      strip.text = element_text(face = 'bold'))
````
