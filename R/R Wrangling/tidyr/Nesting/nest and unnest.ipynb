# Examples

### nest

````r
df <- tibble(x = c(1, 1, 1, 2, 2, 3), y = 1:6, z = 6:1)

df
````

````r
df_nest <- df %>% group_by(x) %>% nest()
df_nest
print(df_nest)
````

````r
#Equivalent
#NOTE: data is the name of the nested column, NOT a keyword argument
df %>% nest(data = c(y, z))
````

````r
#Name the nested column to 'result'
df %>% nest(result = c(y, z))
````

use **tidyselect** syntax and helpers, just like in `dplyr::select()`

````r
df %>% nest(result = any_of(c('y', 'z')))
````

````r
#Each cell is a tibble
iris_nested <- iris %>% nest(petal = starts_with('Petal'), sepal = starts_with('Sepal'))

iris_nested
````

````r
print(iris_nested)
````

````r
iris %>% nest(length = ends_with('Length'), width = ends_with('Width'))
````

### unnest

**`unnest`** is the inverse of **`nest`**

 `unnest()` is primarily designed to work with lists of data frames

````r
df <- tibble(
  x = 1:3,
  y = list(
    NULL,
    tibble(a = 1, b = 2),
    tibble(a = 1:3, b = 3:1)
  )
)

df
````

````r
print(df)
````

````r
#inverse of `nest()`
df %>% unnest(cols = y)
````

````r
df %>% unnest(cols = y, keep_empty = T)
````

You can unnest multiple columns simultaneously

````r
df <- tibble(
 a = list(c("a", "b"), "c"),
 b = list(1:2, 3),
 c = c(11, 22)
)

df
````

````r
print(df)
````

````r
df %>% unnest(c(a, b))
````

Compare with unnesting one column at a time, which generates
the Cartesian product

````r
df
````

````r
df %>% unnest(a)
````

````r
df %>% unnest(a) %>% unnest(b)
````
