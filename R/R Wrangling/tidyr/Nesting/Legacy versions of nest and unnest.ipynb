# Legacy versions of `nest()` and `unnest()`

tidyr 1.0.0 introduced a new syntax for **`nest()`** and **`unnest()`**. 

```r
nest_legacy(data, ..., .key = "data")

unnest_legacy(data, ..., .drop = NA, .id = NULL, .sep = NULL, .preserve = NULL)
```

# Examples

### `nest` and `unnest` are inverse

````r
df <- data.frame(x = c(1, 1, 2), y = 3:1)
df
````

````r
df %>% group_by(x) %>% nest_legacy()
````

````r
#equivalent
df %>% nest_legacy(y)
````

````r
# set the name to 'result'
df %>% nest_legacy(y, .key = 'result')
````

````r
# tidy-select
iris %>% nest_legacy(!Species)
````

````r
df %>% nest_legacy(y) %>% unnest_legacy() 
````

### nesting

````r
iris_nested <- iris %>% nest_legacy(!Species)

iris_nested
````

````r
print(iris_nested)
````

### unnesting

````r
df <- tibble(
  x = 1:2,
  y = list(
    tibble(z = 1),
    tibble(z = 3:4)
  )
)

df
````

````r
print(df)
````

````r
df %>% unnest_legacy(y)
````

You can also unnest multiple columns simultaneously


````r
df <- tibble(
  a = list(c("a", "b"), "c"),
  b = list(1:2, 3),
  c = c(11, 22)
)
````

````r
#Equivalent
#df %>% unnest(c(a, b))
df %>% unnest_legacy(a, b)
````

If you omit the column names, it'll unnest all list-cols

````r
df %>% unnest_legacy()
````
