# Using tidyr in packages


If you know the column names, this code works in the same way regardless of whether its inside or outside of a package:

```r
iris %>% nest(Sepal = c(Sepal.Length, Sepal.Width), Petal = c(Petal.Length, Petal.Width))
```

But `R CMD check` will warn about undefined global variables (Petal.Length, Petal.Width, Sepal.Length, and Sepal.Width), because it doesnâ€™t know that `nest()` is looking for the variables inside of `iris` (i.e. Petal.Length and friends are data-variables, not env-variables).

The easiest way to silence this note is to use `all_of()`

```r
iris %>% nest(Sepal = all_of(c('Sepal.Length', 'Sepal.Width')), Petal = all_of(c('Petal.Length', 'Petal.Width')))
```
