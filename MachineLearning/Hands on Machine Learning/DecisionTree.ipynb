```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV

iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state = 42)
```

```python
grid_params = {
    'max_depth' : [3,4,5],
    'min_samples_split' : [2,3,4],
    'max_leaf_nodes' : [5, 10, 15, 20]
}
grid = GridSearchCV(DecisionTreeClassifier(), grid_params, cv = 5)
grid.fit(X_train, y_train)
```

```python
grid.best_score_
```

```python
grid.best_params_
```

```python
from sklearn.tree import plot_tree
clf = grid.best_estimator_
plt.figure(figsize = (10, 8))
plot_tree(clf, feature_names = iris.feature_names, class_names = iris.target_names)
```

## Estimating Class Probability

```python
clf.predict_proba(X_test)
```

## Regularization Hyperparameters

```python
from sklearn.datasets import make_moons
X, y = make_moons(noise = .1, random_state = 42)
plt.scatter(X[:,0], X[:,1], c = y, cmap = 'Spectral')
```

```python


clf1 = DecisionTreeClassifier()
clf1.fit(X, y)

grid_params = {
    'max_depth' : [3,4,5,6],
    'min_samples_split': [2,3,4],
    'max_leaf_nodes' : [5, 10, 15]
}

grid = GridSearchCV(DecisionTreeClassifier(), grid_params, cv = 5)
grid.fit(X, y)
```

```python
grid.best_score_
```

```python
grid.best_params_
```

```python
def plot(model, ax):
    mglearn.discrete_scatter(X[:,0], X[:,1], y, ax = ax)
    mglearn.plots.plot_2d_separator(model, X, ax = ax)
    mglearn.plots.plot_2d_classification(model, X, alpha = .3, ax = ax)

clf2 = grid.best_estimator_
fig, axes = plt.subplots(1,2,subplot_kw = {'xticks' : [], 'yticks' : []})
plot(clf1, axes[0])
plot(clf2, axes[1])
axes[0].set_title('Overfitting')
axes[1].set_title('Pre-pruning')
    
```

## Regression

```python
N = 100
np.random.seed(6)
X = np.linspace(-2, 2, N)
y = X * X + 2 * np.random.rand(N)
plt.scatter(X, y)
```

```python
from sklearn.tree import DecisionTreeRegressor
clf1 = DecisionTreeRegressor()
XX = X[:,None]
clf1.fit(XX, y)

params = {
    'max_depth' : [2,3,4,5],
    'min_samples_split' : [2,3,4,5,6]
}
grid = GridSearchCV(DecisionTreeRegressor(), params, cv = 5, scoring = 'neg_mean_squared_error')
grid.fit(XX, y)
```

```python
def plot_prediction(model, ax):
    ax.scatter(X, y, label = 'True value',s  = 10)
    y_pred = model.predict(XX)
    ax.plot(X, y_pred, color = 'red', label = 'prediction')
    ax.set_title('max_depth = %d\nmin_samples_split = %d' % (model.get_depth(), model.min_samples_split), fontsize = 7)
    ax.legend()
```

```python
plt.style.use('default')
fig, axes = plt.subplots(1,2, sharey = True, figsize = (5, 3))
clf2 = grid.best_estimator_
plot_prediction(clf1, axes[0])
plot_prediction(clf2, axes[1])
```

```python
plot_tree(clf2)
```

## Presort

<p>
Presorting the training set speeds up training only if the dataset is <u>smaller than a
    few thousand</u> instances. If it contains 100,000 instances, setting <code>presort=True</code>
will considerably slow down training.
